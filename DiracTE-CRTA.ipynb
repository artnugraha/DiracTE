# Constant Relaxation Time Approximation

Using constant relaxation time approximation, the transport equation will become $\mathcal{T}(E)=v_x^2C\mathcal{D}(E)$ with longitudinal velocity $v_x$, and density of states $\mathcal{D}$. The density of states per unit area $\mathcal{D}(E)=d(E)/A$ is defined by

\begin{align}
    d(E)&=\sum_{\textbf{k}}\delta(E_\textbf{k}-E) \\
    \mathcal{D}(E)&=\frac{1}{(2\pi)^2}\int d^2\textbf{k}\delta(E_\textbf{k}-E), \quad E_\textbf{k}^2=(\hbar vk)^2+\Delta^2\\
    &=\frac{1}{(2\pi)^2}2\pi\int k\,dk\,\delta(E_\textbf{k}E),\quad (\hbar v)^2k\,dk=E_\textbf{k}dE_\textbf{k} \\
    &=\frac{1}{2\pi(\hbar v)^2}\int_{-\infty}^{\infty}E_\textbf{k}dE_\textbf{k}\delta(E_\textbf{k}-E) \\
    &=\frac{|E|}{2\pi(\hbar v)^2}\Theta(|E|-|\Delta|),
\end{align}
where $\Theta(x)$ is a step function, $\Theta(x)=1$ if $x>0$ and $\Theta(x)=0$ otherwise.


# Gapless

## Gapless one band model
With $v_x^2=v^2/d$, for conduction band,
\begin{align}
    \mathcal{L}_i^+&=\frac{Cv^2}{d}\frac{1}{2\pi (\hbar v)^2}\int_0^\infty dE\,E(E-\mu)^i\frac{1}{k_BT}\frac{\exp(\frac{E-\mu}{k_BT})}{(\exp(\frac{E-\mu}{k_BT})+1)^2}, \\
    &=\frac{C}{2d\pi\hbar^2}(k_BT)^{i+1}\int_0^\infty d\varepsilon\,\varepsilon(\varepsilon-\eta)^i\frac{\exp(\varepsilon-\eta)}{(\exp(\varepsilon-\eta)+1)^2}, \\
    &=\frac{C}{2d\pi\hbar^2}(k_BT)^{i+1}\int_{-\eta}^\infty dx\,x^i(x+\eta)\frac{\exp(x)}{(\exp(x)+1)^2}, \\
    &=\frac{C}{2d\pi\hbar^2}(k_BT)^{i+1}\left(\mathcal{F}_{i+1}^+(\eta)+\eta\mathcal{F}_i^+(\eta)\right),
\end{align}

where $E=\varepsilon\,k_BT$, $\mu=\eta\,k_BT$, and $\mathcal{F}_i^+(\eta)=\int_{-\eta}^\infty dx\,x^i\frac{\exp(x)}{(\exp(x)+1)^2}$. 

For $i = 0,1,2$, we have:
\begin{align}
\mathcal{F}_0^+ (\eta) &= \frac{e^\eta}{e^\eta + 1},\\
\mathcal{F}_1^+ (\eta) &= \frac{\eta}{e^\eta + 1} + \ln(1 + e^{-\eta}),\\
\mathcal{F}_2^+ (\eta) &= \frac{\pi^2}{3} - \frac{\eta^2}{e^\eta + 1}
                            - 2\eta \ln(1 + e^{-\eta}) + 2 \mathrm{Li}_2(-e^{-\eta}),\\
\mathcal{F}_3^+ (\eta) &= \eta^2\left(\frac{\eta}{1+e^\eta}+3\ln\left(1+e^{-\eta}\right)\right)-6\eta\mathrm{Li}_2(-e^{-\eta})-6\mathrm{Li}_3(-e^{-\eta})
\end{align}
where $\mathrm{Li}_k(z) = \displaystyle\sum_{n=1}^\infty \frac{z^n}{n^k}$ is the polylogarithmic function.


Considering only the conduction band, the transport properties can be written as follows:

\begin{align}
    \sigma^+&=q^2\mathcal{L}_0^+ \\
        &=\frac{Cq^2k_BT}{2d\pi\hbar^2}\left(\mathcal{F}_1^+(\eta)+\eta\mathcal{F}_0^+(\eta)\right) \\
        &\equiv\frac{Cq^2k_BT}{2d\pi\hbar^2}\tilde{\sigma}^+,\\
\\
    S^+&=-\frac{1}{qT}\frac{\mathcal{L}_1^+}{\mathcal{L}_0^+} \\
       &=-\frac{k_B}{q}\frac{\left(\mathcal{F}_{2}^+(\eta)+\eta\mathcal{F}_1^+(\eta)\right)}{\left(\mathcal{F}_{1}^+(\eta)+\eta\mathcal{F}_0^+(\eta)\right)} \\
       &\equiv-\frac{k_B}{q}\tilde{S}^+,\\
\\
    \kappa_e^+ &= \frac{1}{T} \left(\mathcal{L}_2^+ -
\frac{(\mathcal{L}_1^+)^2}{\mathcal{L}_0^+}\right)\\
        &=\frac{Ck_B^3T^2}{2d\pi\hbar^2}
            \left(\left(\mathcal{F}_{3}^+(\eta)+\eta\mathcal{F}_2^+(\eta)\right)-\frac{\left(\mathcal{F}_{2}^+(\eta)+\eta\mathcal{F}_1^+(\eta)\right)^2}{\left(\mathcal{F}_{1}^+(\eta)+\eta\mathcal{F}_0^+(\eta)\right)}
            \right) \\
        &=\kappa_0
            \left(\left(\mathcal{F}_{3}^+(\eta)+\eta\mathcal{F}_2^+(\eta)\right)-\frac{\left(\mathcal{F}_{2}^+(\eta)+\eta\mathcal{F}_1^+(\eta)\right)^2}{\left(\mathcal{F}_{1}^+(\eta)+\eta\mathcal{F}_0^+(\eta)\right)}
            \right) \\
        &\equiv\kappa_0\tilde{\kappa_e}^+,
\\
\end{align}
where $\kappa_0=\frac{Ck_B^3T^2}{2d\pi\hbar^2}$. The phonon thermal conductiviy can also be represented as,
    \begin{align}
    \kappa_{ph}=r_\kappa\kappa_0.
    \end{align}
    
Based on those transport quantities, the thermoelectric figure of merit ($ZT$) can be calculated as,
    \begin{align}
    ZT&=\frac{(S^+)^2\sigma^+}{\kappa_e^+ + \kappa_{ph}}T \\
    &=\frac{(\tilde{S}^+)^2\tilde{\sigma}^+}{\tilde{\kappa}_e^+ + r_\kappa}
    \end{align}

### Code: Gapless one band

```python
###################################################
# ZT of GAPLESS 2D Dirac material: ONE BAND model #
#            Constant Relaxation                  #
###################################################

#----------#
# PREAMBLE #
#----------#

# Libraries below will be used throughout the notebook.
# Therefore, we must execute this cell at least once.

# Call necessary mathematical and numerical libraries
from numpy import frompyfunc, linspace, exp, log, pi, zeros
from mpmath import polylog
from scipy.integrate import quad

# Call plotting library
import matplotlib.pyplot as plt
%matplotlib inline

SMALL_SIZE = 12
MEDIUM_SIZE = 13
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)        # controls default text sizes
plt.rc('legend', fontsize=SMALL_SIZE)  # legend fontsize
plt.rc('xtick', labelsize=MEDIUM_SIZE) # fontsize of the x tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE) # fontsize of the y tick labels
plt.rc('axes', labelsize=BIGGER_SIZE)  # fontsize of the x and y labels

# Specific treatment for polylogarithm function
plog = frompyfunc(polylog, 2, 1)
# numpy.frompyfunc(func, nin, nout)
# Takes an arbitrary Python function and returns a NumPy ufunc.
# Parameters:
#            func : An arbitrary Python function.
#            nin  : (int) The number of input arguments.
#            nout : (int) The number of objects returned by func.
# Returns a NumPy universal function (ufunc) object.

#-----------#
# MAIN PART #
#-----------#

# Define kernel:
def F0(x):
    return exp(x)/(1+exp(x))
def F1(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def F3(x):
    return ( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x)) - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )

# ZT function:
def ZT(Seeb,Sigma,KappaE,KappaP):
    return (Seeb**2 * Sigma)/(KappaE + KappaP)

# Create sample points within [-10,10]:
npoint = 101
eta = linspace(-10, 10, npoint) 

# TE quantities:
S_c = (F2(eta)+eta*F1(eta)) / (F1(eta)+eta*F0(eta))
Sigma_c = F1(eta)+eta*F0(eta)
kappa_c = (F3(eta)+eta*F2(eta)) - ( (F2(eta)+eta*F1(eta))**2 / (F1(eta)+eta*F0(eta)) )

#----------#
# PLOTTING #
#----------#

# Example line formats are:
# 'b-'  = blue solid line
# 'r--' = red dashed line
# 'g:   = green dotted line
# if black is needed, use 'k' option (as in cmyk color)

# ---ZT----
plt.figure(figsize=(4,3))
plt.plot(eta, ZT(S_c, Sigma_c, kappa_c, 0), 'b-',
         eta, ZT(S_c, Sigma_c, kappa_c, 1), 'r--',
         eta, ZT(S_c, Sigma_c, kappa_c, 2), 'g:')
plt.axis([-4,4,0,4]) # the format is [xmin,xmax,ymin,ymax]
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-2.1,3.5, r'$\alpha = 0$', color='b')
plt.text(-2.1,2, r'$\alpha = 1$', color='r')
plt.text(-2.1,0.6, r'$\alpha = 2$', color='g')
# in plt.xlabel or ylabel, the option r is not necessary
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$ZT$')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()

# Save figure in PDF file
plt.savefig('CRT-ZTgapless1band.pdf')

# ----Transport Quantities----
plt.figure(figsize=(3.5,3.5))
plt.plot(eta, S_c, 'b-',
         eta, Sigma_c, 'r--',
         eta, kappa_c, 'g:')
plt.axis([-4,4,0,5]) # the format is [xmin,xmax,ymin,ymax]
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-1,2.5, r'$S$', color='b')
plt.text(2.5,2.4, r'$\sigma$', color='r')
plt.text(-1,1, '$\kappa_e$', color='g')
# in plt.xlabel or ylabel, the option r is not necessary
plt.xlabel('$\mu/k_B T$')
plt.ylabel('Transport quantities')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()

# Save figure in PDF file
plt.savefig('CRT-TQGapless1band.pdf')
```

## Gapless two bands model

Here we consider both conduction and valence bands are contributing to the transport phenomena.

For valence band,

\begin{align}
    \mathcal{L}^-_i&=\frac{Cv^2}{d}\frac{1}{2\pi(\hbar v)^2}\int_{-\infty}^0dE\,(-E)(E-\mu)^i\frac{1}{k_BT}\frac{\exp(\frac{E-\mu}{k_BT})}{(\exp(\frac{E-\mu}{k_BT})+1)^2},\\
        &=-\frac{C}{2d\pi\hbar^2}(k_BT)^{i+1}\int_{-\infty}^0 d\varepsilon\,\varepsilon(\varepsilon-\eta)^i\frac{\exp(\varepsilon-\eta)}{(\exp(\varepsilon-\eta)+1)^2},\\
        &=-\frac{C}{2d\pi\hbar^2}(k_BT)^{i+1}\int_{-\infty}^{-\eta} dx\,x^i(x+\eta)\frac{\exp(x)}{(\exp(x)+1)^2}, \\
        &=-\frac{C}{2d\pi\hbar^2}(k_BT)^{i+1}\left(\mathcal{F}_{i+1}^-(\eta)+\eta\mathcal{F}_i^-(\eta)\right),
\end{align}
where $\mathcal{F}_i^-(\eta)=\int_{-\infty}^{-\eta} dx\,x^i\frac{\exp(x)}{(\exp(x)+1)^2}$. For $i = 0,1,2$, we have:
\begin{align}
\mathcal{F}_{0}^- (\eta) &= \frac{1}{e^\eta + 1},\\
\mathcal{F}_{1}^- (\eta) &= - \frac{\eta}{e^\eta + 1} - \ln(1 + e^{-\eta}),\\
\mathcal{F}_{2}^- (\eta) &= \frac{\eta^2}{e^\eta + 1}
                            + 2\eta \ln(1 + e^{-\eta}) - 2 \mathrm{Li}_2(-e^{-\eta}), \\
\mathcal{F}_3^-(\eta)&=\eta^2\left(-\frac{\eta}{1+e^\eta}-3\ln(1+e^{-\eta})\right)+6\eta\mathrm{Li}_2(-e^{-\eta})+6\mathrm{Li}_3(-e^{-\eta})
\end{align}

[comment]: <> (Notice that the integrals $\mathcal{F}_{i,c}$ and $\mathcal{F}_{i,v}$ are connected via electron-hole symmetry       of the system:
\begin{align}
\mathcal{F}_0^+ (\eta) &= \mathcal{F}_0^- (-\eta)\\
\mathcal{F}_1^+ (\eta) &= -\mathcal{F}_1^- (\eta)\\
\mathcal{F}_2^+ (\eta) &= \mathcal{F}_2^-(-\eta)\\
\mathcal{F}_3^+ (\eta) &= -\mathcal{F}_3^-(\eta)
\end{align})

The contribution of the valence band to the TE quantities, i.e., $\sigma^-$, $S^-$, and $\kappa_e^-$, can be expressed similarly with that of the conduction band:
\begin{align}
    \sigma^-&=-\frac{Cq^2k_BT}{2d\pi\hbar^2}\left(\mathcal{F}_1^-(\eta)+\eta\mathcal{F}_0^-(\eta)\right) \\
        &\equiv\frac{Cq^2k_BT}{2d\pi\hbar^2}\tilde{\sigma}^-,\\
\\
    S^-&=\frac{k_B}{q}\frac{\left(\mathcal{F}_{2}^-(\eta)+\eta\mathcal{F}_1^-(\eta)\right)}{\left(\mathcal{F}_{1}^-(\eta)+\eta\mathcal{F}_0^-(\eta)\right)} \\
       &\equiv\frac{k_B}{q}\tilde{S}^-,\\
\\
    \kappa_e^-&=-\kappa_0
            \left(\left(\mathcal{F}_{3}^-(\eta)+\eta\mathcal{F}_2^-(\eta)\right)-\frac{\left(\mathcal{F}_{2}^-(\eta)+
            \eta\mathcal{F}_1^-(\eta)\right)^2}{\left(\mathcal{F}_{1}^-(\eta)+\eta\mathcal{F}_0^-(\eta)\right)}
            \right) \\
        &\equiv-\kappa_0\tilde{\kappa_e}^-,
\\
\end{align}
Then the thermoelectric quantities itself can be calculated by,
\begin{align}
\sigma &= \sigma^+ + \sigma^-\\
S &= \frac{\sigma^+ S^+ + \sigma^- S^-}{\sigma^+ + \sigma^-}\\
\kappa_e &= \frac{\sigma^+ \sigma^-}{\sigma^+ + \sigma^-}(S^+ - S^-)^2
            + (\kappa_e^+ + \kappa_e^-)
\end{align}

### Code: Gapless two bands

```python
###################################################
# ZT of GAPLESS 2D Dirac material: TWO BAND model #
#           Constant Relaxation Time              #
###################################################

#----------#
# PREAMBLE #
#----------#

# We use the same preamble as in the one band model.
# Make sure the one band part has been executed! 
# (CTRL + ENTER) in that cell!

#-----------#
# MAIN PART #
#-----------#

# Define kernel:
## conduction band
def F0c(x):
    return exp(x)/(1+exp(x))
def F1c(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2c(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def F3c(x):
    return ( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x)) - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )

## valence band
def F0v(x):
    return 1/(1+exp(x))
def F1v(x):
    return ( -x/(1+exp(x)) - log(1+exp(-x)) )
def F2v(x):
    return ( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )
def F3v(x):
    return ( -(x**3)/(1+exp(x)) - 3*(x**2)*log(1+exp(-x)) + 6*x*plog(2,-exp(-x)) + 6*plog(3,-exp(-x)) )

# ZT function:
def ZT(Seeb,Sigma,KappaE,KappaP):
    return (Seeb**2 * Sigma)/(KappaE + KappaP)

# Create sample points within [-10,10]:
npoint = 101
eta = linspace(-10, 10, npoint) 

# TE quantities:
S_c = -( F2c(eta) + eta*F1c(eta) ) / ( F1c(eta) + eta*F0c(eta) )
Sigma_c = F1c(eta) + eta*F0c(eta)
kappa_c = ( F3c(eta) + eta*F2c(eta) ) - ( (F2c(eta)+eta*F1c(eta))**2 / (F1c(eta)+eta*F0c(eta)) )

S_v = -(F2v(eta)+eta*F1v(eta)) / (F1v(eta)+eta*F0v(eta))
Sigma_v = -F1v(eta)-eta*F0v(eta)
kappa_v = -(F3v(eta)+eta*F2v(eta)) + ( (F2v(eta)+eta*F1v(eta))**2 / (F1v(eta)+eta*F0v(eta)) )

Sigma = Sigma_c + Sigma_v
S = (Sigma_c * S_c + Sigma_v * S_v) / Sigma
kappae = ( (kappa_c + kappa_v)
          + (Sigma_c * Sigma_v)*(S_c - S_v)**2 / (Sigma) )

#----------#
# PLOTTING #
#----------#

# Example line formats are:
# 'b-'  = blue solid line
# 'r--' = red dashed line
# 'g:   = green dotted line
# if black is needed, use 'k' option (as in cmyk color)

# ---ZT---
plt.figure(figsize=(4,4))
plt.plot(eta, ZT(S, Sigma, kappae, 0), 'r-',
         eta, ZT(S, Sigma, kappae, 1), 'b-',
         eta, ZT(S, Sigma, kappae, 2), 'g-')
plt.axis([-4,4,0,0.3]) # the format is [xmin,xmax,ymin,ymax]
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$ZT$')
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-1,0.3, r'$\alpha=0$', color='r')
plt.text(-1,0.28, r'$\alpha=1$', color='b')
plt.text(-1,0.26, r'$\alpha=2$', color='g')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()

# Save figure in PDF file
plt.savefig('ZTgapless2band.pdf')

# ---Transport Quantities---
plt.figure(figsize=(4,4))
plt.plot(eta, S, 'b-',
         eta, Sigma, 'r--',
         eta, kappae, 'g:')
plt.axis([-4,4,-2,15]) # the format is [xmin,xmax,ymin,ymax]
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-1,0, r'$S$', color='b')
plt.text(-1,2.5, r'$\sigma$', color='r')
plt.text(-1,9, r'$\kappa_e$', color='g')
# in plt.xlabel or ylabel, the option r is not necessary
plt.xlabel('$\mu/k_B T$')
plt.ylabel('Transport quantities')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()

# Save figure in PDF file
plt.savefig('TQGapless2band.pdf')

# ---Showing Particle-Hole Symmetry---
plt.figure(figsize=(3.5,3.5))
plt.plot(eta, S_c, 'b-',
         eta, S_v, 'r-',
         eta, S, 'g:')
plt.axis([-4,4,-4,4]) # the format is [xmin,xmax,ymin,ymax]
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-1,-2, r'$S_c$', color='b')
plt.text(-1,2.5, r'$S_v$', color='r')
plt.text(-1,1, r'$S$', color='g')
# in plt.xlabel or ylabel, the option r is not necessary
plt.xlabel('$\mu/k_B T$')
plt.ylabel('Transport quantities')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()
# Save figure in PDF file
plt.savefig('SGapless.pdf')

plt.figure(figsize=(3.5,3.5))
plt.plot(eta, Sigma_c, 'b-',
         eta, Sigma_v, 'r-',
         eta, Sigma, 'g:')
plt.axis([-4,4,0,4]) # the format is [xmin,xmax,ymin,ymax]
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-1,0.1, r'$\sigma_c$', color='b')
plt.text(-1,0.8, r'$\sigma_v$', color='r')
plt.text(-1,2, '$\sigma$', color='g')
# in plt.xlabel or ylabel, the option r is not necessary
plt.xlabel('$\mu/k_B T$')
plt.ylabel('Transport quantities')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()
# Save figure in PDF file
plt.savefig('SigGapless.pdf')

plt.figure(figsize=(3.5,3.5))
plt.plot(eta, kappa_c, 'b-',
         eta, kappa_v, 'r-',
         eta, kappae, 'g:')
plt.axis([-4,4,0,15]) # the format is [xmin,xmax,ymin,ymax]
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-2.5,1.4, r'$\kappa_{e,c}$', color='b')
plt.text(2,1.4, r'$\kappa_{e,v}$', color='r')
plt.text(0,9, r'$\kappa_e$', color='g')
# in plt.xlabel or ylabel, the option r is not necessary
plt.xlabel('$\mu/k_B T$')
plt.ylabel('Transport quantities')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()
# Save figure in PDF file
plt.savefig('KapGapless.pdf')
```

# Gapped

For gapped 2D Dirac material (take 2-dimension; $d=2$), the longitudinal velocity is given by,
    \begin{align}
    v_x=\pm\frac{v^2}{2}\left(\frac{E^2-\Delta^2}{E^2}\right).
    \end{align}
    
[comment]: <> (Hence, the Transport Distribution Function for this case can be written as,
    \begin{align}
    \mathcal{T}=C\frac{E}{2\pi(\hbar v)^2}\Theta(|E|-|\Delta|)
    \end{align})
    
## Gapped one band model

For the conduction band,
    \begin{align}
    \mathcal{L}_i^+&=\frac{Cv^2}{2}\frac{1}{2\pi(\hbar v)^2}\int_\Delta^\infty dE\,\left(\frac{E^2-\Delta^2}{E^2}\right) E (E-\mu)^i\frac{1}{k_BT}\frac{\exp(\frac{E-\mu}{k_BT})}{(\exp(\frac{E-\mu}{k_BT})+1)^2}, \\
    &=\frac{C}{4\pi\hbar^2}(k_BT)^{i+1}\int_\tilde{\Delta}^\infty d\varepsilon\,\left(\frac{\varepsilon^2-\tilde{\Delta}^2}{\varepsilon^2}\right) \varepsilon (\varepsilon-\eta)^i\frac{\exp(\varepsilon-\eta)}{(\exp(\varepsilon-\eta)+1)^2},\\
    &=\frac{C}{4\pi\hbar^2}(k_BT)^{i+1}\int_{\tilde{\Delta}-\eta}^\infty dx\,\left(1-\frac{\tilde{\Delta}^2}{(x+\eta)^2}\right)x^i(x+\eta)\frac{\exp(x)}{(\exp(x)+1)^2}, \\
    &=\frac{C}{4\pi\hbar^2}(k_BT)^{i+1}\left(\mathcal{F}_{i+1}^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_i^+(\eta-\tilde{\Delta})-\tilde{\mathcal{G}}_i^+(\tilde{\Delta},\eta)\right),
    \end{align}
with
    \begin{align}
\tilde{\mathcal{G}}_i^+(\tilde{\Delta},\eta)
= \int_{\tilde\Delta-\eta}^\infty \frac{\tilde\Delta^2}{(x+\eta)}
\frac{x^i e^x}{(e^x+1)^2} dx.
    \end{align}
This $\tilde{\mathcal{G}}_i$ integral will be calculated numerically because it is difficult to solve the integral analytically.


Then, the TE quantities will be,

\begin{align}
    \sigma^+&=q^2\mathcal{L}_0^+ \\
        &=\frac{Cq^2k_BT}{2d\pi\hbar^2}\left(\mathcal{F}_1^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_0^+(\eta-\tilde{\Delta})-\mathcal{G}_0^+(\tilde{\Delta},\eta)\right) \\
        &\equiv\frac{Cq^2k_BT}{2d\pi\hbar^2}\tilde{\sigma}^+,\\
\\
    S^+&=-\frac{1}{qT}\frac{\mathcal{L}_1^+}{\mathcal{L}_0^+} \\
       &=-\frac{k_B}{q}\frac{\left(\mathcal{F}_{2}^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_1^+(\eta-\tilde{\Delta})-\mathcal{G}_1^+(\tilde{\Delta},\eta)\right)}{\left(\mathcal{F}_{1}^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_0^+(\eta-\tilde{\Delta})-\mathcal{G}_0^+(\tilde{\Delta},\eta)\right)} \\
       &\equiv-\frac{k_B}{q}\tilde{S}^+,\\
\\
    \kappa_e^+ &= \frac{1}{T} \left(\mathcal{L}_2^+ -
\frac{(\mathcal{L}_1^+)^2}{\mathcal{L}_0^+}\right)\\
        &=\frac{Ck_B^3T^2}{2d\pi\hbar^2}
            \left(\left(\mathcal{F}_{3}^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_2^+(\eta-\tilde{\Delta})-\mathcal{G}_2^+(\tilde{\Delta},\eta)\right)
            -\frac{\left(\mathcal{F}_{2}^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_1^+(\eta-\tilde{\Delta})-\mathcal{G}_1^+(\tilde{\Delta},\eta)\right)^2}{\left(\mathcal{F}_{1}^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_0^+(\eta-\tilde{\Delta})-\mathcal{G}_0^+(\tilde{\Delta},\eta)\right)}
            \right) \\
        &=\kappa_0
            \left(\left(\mathcal{F}_{3}^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_2^+(\eta-\tilde{\Delta})-\mathcal{G}_2^+(\tilde{\Delta},\eta)\right)
            -\frac{\left(\mathcal{F}_{2}^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_1^+(\eta-\tilde{\Delta})-\mathcal{G}_1^+(\tilde{\Delta},\eta)\right)^2}{\left(\mathcal{F}_{1}^+(\eta-\tilde{\Delta})+\eta\mathcal{F}_0^+(\eta-\tilde{\Delta})-\mathcal{G}_0^+(\tilde{\Delta},\eta)\right)}
            \right) \\
        &\equiv\kappa_0\tilde{\kappa_e}^+,
\\
\end{align}
where $\kappa_0=\frac{Ck_B^3T^2}{2d\pi\hbar^2}$. The phonon thermal conductiviy can also be represented as,
    \begin{align}
    \kappa_{ph}=r_\kappa\kappa_0.
    \end{align}
    
Based on those transport quantities, the thermoelectric figure of merit ($ZT$) can be calculated as,
    \begin{align}
    ZT&=\frac{(S^+)^2\sigma^+}{\kappa_e^+ + \kappa_{ph}}T \\
    &=\frac{(\tilde{S}^+)^2\tilde{\sigma}^+}{\tilde{\kappa}_e^+ + r_\kappa}
    \end{align}

### Code: Gapped one band model

```python
##################################################
# ZT of GAPPED 2D Dirac material: ONE BAND model #
#                     CRT                        #
##################################################

#-----------#
# MAIN PART #
#-----------#

# Define kernel:

delta    = 1    # band gap
infinity = 100  # this value is sufficient for our function

def F0(x):
    return exp(x)/(1+exp(x))
def F1(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def F3(x):
    return ( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x)) - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )

def G0(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I
def G1(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I
def G2(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I

# ZT function:
def ZT(Seeb,Sig,KappaE,KappaP):
    return (Seeb**2 * Sig)/(KappaE + KappaP)

# Create sample points within [-10,10]:
npoint  = 101
eta     = linspace(-10, 10, npoint) 

L0      = zeros(npoint,float)
L1      = zeros(npoint,float)
L2      = zeros(npoint,float)

S_c     = zeros(npoint,float)
Sigma_c = zeros(npoint,float)
kappa_c = zeros(npoint,float)

for i in range(eta.size):
    L0[i] = F1(eta[i] - delta) + eta[i] * F0(eta[i] - delta) - delta**2 * G0(eta[i],delta-eta[i])
    L1[i] = F2(eta[i] - delta) + eta[i] * F1(eta[i] - delta) - delta**2 * G1(eta[i],delta-eta[i])
    L2[i] = F3(eta[i] - delta) + eta[i] * F2(eta[i] - delta) - delta**2 * G2(eta[i],delta-eta[i])

# TE quantities:
S_c     = -L1 / L0
Sigma_c =  L0
kappa_c =  L2 - (L1**2 / L0)
    
#----------#
# PLOTTING #
#----------#

plt.figure(figsize=(4,3))
plt.plot(eta, ZT(S_c, Sigma_c, kappa_c, 0), 'b-',
         eta, ZT(S_c, Sigma_c, kappa_c, 1), 'r--',
         eta, ZT(S_c, Sigma_c, kappa_c, 2), 'g:')
plt.axis([-4,4,0,4]) # the format is [xmin,xmax,ymin,ymax]
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-0.7,3.0, r'$r_\kappa = 0$', color='b')
plt.text(-0.7,2.5, r'$r_\kappa = 1$', color='r')
plt.text(-0.7,0.8, r'$r_\kappa = 2$', color='g')
# in plt.xlabel or ylabel, the option r is not necessary
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$ZT$')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()

# Save figure in PDF file
plt.savefig('ZTgapped1band.pdf')
```

## Gapped two bands model

For valence band,
\begin{align}
    \mathcal{L}_i^-&=\left(\frac{Cv^2}{2}\right)\left(-\frac{1}{2\pi(\hbar v)^2}\right)\int_{-\infty}^{-\Delta} dE\,\left(\frac{E^2-\Delta^2}{E^2}\right) E (E-\mu)^i\frac{1}{k_BT}\frac{\exp(\frac{E-\mu}{k_BT})}{(\exp(\frac{E-\mu}{k_BT})+1)^2}, \\
    &=-\frac{C}{4\pi\hbar^2}(k_BT)^{i+1}\int_{-\infty}^{-\tilde{\Delta}} d\varepsilon\,\left(\frac{\varepsilon^2-\tilde{\Delta}^2}{\varepsilon^2}\right) \varepsilon (\varepsilon-\eta)^i\frac{\exp(\varepsilon-\eta)}{(\exp(\varepsilon-\eta)+1)^2},\\
    &=-\frac{C}{4\pi\hbar^2}(k_BT)^{i+1}\int^{-\tilde{\Delta}-\eta}_{-\infty} dx\,\left(1-\frac{\tilde{\Delta}^2}{(x+\eta)^2}\right)x^i(x+\eta)\frac{\exp(x)}{(\exp(x)+1)^2}, \\
    &=-\frac{C}{4\pi\hbar^2}(k_BT)^{i+1}\left(\mathcal{F}_{i+1}^-(\eta+\tilde{\Delta})+\eta\mathcal{F}_i^-(\eta+\tilde{\Delta})-\tilde{\mathcal{G}}_i^-(\tilde{\Delta},\eta)\right),
    \end{align}
with
    \begin{align}
\tilde{\mathcal{G}}_i^-(\tilde{\Delta},\eta)
= \int^{-\tilde{\Delta}-\eta}_{-\infty}\frac{\tilde\Delta^2}{(x+\eta)}
\frac{x^i e^x}{(e^x+1)^2} dx.
    \end{align}

Here we consider both conduction and valence bands are contributing to the transport phenomena.

Then the thermoelectric quantities itself can be calculated by,
\begin{align}
\sigma &= \sigma^+ + \sigma^-\\
S &= \frac{\sigma^+ S^+ + \sigma^- S^-}{\sigma^+ + \sigma^-}\\
\kappa_e &= \frac{\sigma^+ \sigma^-}{\sigma^+ + \sigma^-}(S^+ - S^-)^2
            + (\kappa_e^+ + \kappa_e^-)
\end{align}

### Code: Gapped two bands

```python
##################################################
# ZT of GAPPED 2D Dirac material: TWO BAND model #
#                       CRT                      #
##################################################

#-----------#
# MAIN PART #
#-----------#

# Define kernel:

delta    = 1    # band gap
infinity = 100  # this value is sufficient for our function

## conduction band
def F0c(x):
    return exp(x)/(1+exp(x))
def F1c(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2c(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def F3c(x):
    return ( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x)) - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )

def G0c(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I
def G1c(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I
def G2c(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I

## valence band
def F0v(x):
    return 1/(1+exp(x))
def F1v(x):
    return ( -x/(1+exp(x)) - log(1+exp(-x)) )
def F2v(x):
    return ( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )
def F3v(x):
    return ( -(x**3)/(1+exp(x)) - 3*(x**2)*log(1+exp(-x)) + 6*x*plog(2,-exp(-x)) + 6*plog(3,-exp(-x)) )

def G0v(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D, args=(xx))
    return I
def G1v(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D, args=(xx))
    return I
def G2v(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D, args=(xx))
    return I

# ZT function:
def ZT(Seeb,Sig,KappaE,KappaP):
    return (Seeb**2 * Sigma)/(KappaE + KappaP)

# Create sample points within [-10,10]:
npoint  = 101
eta     = linspace(-10, 10, npoint) 

L0c     = zeros(npoint,float)
L1c     = zeros(npoint,float)
L2c     = zeros(npoint,float)

L0v     = zeros(npoint,float)
L1v     = zeros(npoint,float)
L2v     = zeros(npoint,float)

S_c     = zeros(npoint,float)
Sigma_c = zeros(npoint,float)
kappa_c = zeros(npoint,float)

S_v     = zeros(npoint,float)
Sigma_v = zeros(npoint,float)
kappa_v = zeros(npoint,float)

S       = zeros(npoint,float)
Sigma   = zeros(npoint,float)
kappae  = zeros(npoint,float)

for i in range(eta.size):
    L0c[i] = F1c(eta[i] - delta) + eta[i] * F0c(eta[i] - delta) - delta**2 * G0c(eta[i],delta-eta[i])
    L1c[i] = F2c(eta[i] - delta) + eta[i] * F1c(eta[i] - delta) - delta**2 * G1c(eta[i],delta-eta[i])
    L2c[i] = F3c(eta[i] - delta) + eta[i] * F2c(eta[i] - delta) - delta**2 * G2c(eta[i],delta-eta[i])
    L0v[i] = -1 * (F1v(eta[i] + delta) + eta[i] * F0v(eta[i] + delta) - delta**2 * G0v(eta[i],delta+eta[i]))
    L1v[i] = -1 * (F2v(eta[i] + delta) + eta[i] * F1v(eta[i] + delta) - delta**2 * G1v(eta[i],delta+eta[i]))
    L2v[i] = -1 * (F3v(eta[i] + delta) + eta[i] * F2v(eta[i] + delta) - delta**2 * G2v(eta[i],delta+eta[i]))   
    
# TE quantities:
S_c     = -L1c / L0c
S_v     = -L1v / L0v
Sigma_c =  L0c
Sigma_v =  L0v
kappa_c =  L2c - (L1c**2 / L0c)
kappa_v =  L2v - (L1v**2 / L0v)

Sigma   =  Sigma_c + Sigma_v
S       = (Sigma_c * S_c + Sigma_v * S_v) / Sigma
kappae  = ( (kappa_c + kappa_v)
          + (Sigma_c * Sigma_v)*(S_c - S_v)**2 / (Sigma) )

#----------#
# PLOTTING #
#----------#

# Example line formats are:
# 'b-'  = blue solid line
# 'r--' = red dashed line
# 'g:   = green dotted line
# if black is needed, use 'k' option (as in cmyk color)

#-----------------------#
# SEEBECK'S COEFFICIENT #
#-----------------------#

plt.figure(figsize=(4,3))
plt.plot(eta, S_c, 'r--',
         eta, S_v, 'g:',
         eta, S, 'b-')
plt.axis([-8,8,-4,4]) # the format is [xmin,xmax,ymin,ymax]
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$S/S_0$')
plt.tight_layout()

# Save figure in PDF file
plt.savefig('gappedSeebeck.pdf')

#-------------------------#
# ELECTRICAL CONDUCTIVITY #
#-------------------------#

plt.figure(figsize=(4,3))
plt.plot(eta, Sigma_c, 'r--',
         eta, Sigma_v, 'g:',
         eta, Sigma, 'b-')
plt.axis([-8,8,0,1]) # the format is [xmin,xmax,ymin,ymax]
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$\sigma/\sigma_0$')
plt.tight_layout()

# Save figure in PDF file
plt.savefig('gappedSigma.pdf')

#-----------------#
# FIGURE OF MERIT #
#-----------------#

plt.figure(figsize=(4,3))
plt.plot(eta, ZT(S, Sigma, kappae, 0), 'b-',
         eta, ZT(S, Sigma, kappae, 1), 'r--',
         eta, ZT(S, Sigma, kappae, 2), 'g:')
plt.axis([-8,8,0,0.6]) # the format is [xmin,xmax,ymin,ymax]
plt.text(4,0.4, r'$r_\kappa = 0$', color='b')
plt.text(4,0.35, r'$r_\kappa = 1$', color='r')
plt.text(4,0.3, r'$r_\kappa = 2$', color='g')
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$ZT$')
plt.tight_layout()

# Save figure in PDF file
plt.savefig('ZTgapped2band.pdf')
```

```python
##################################
# TE of GAPPED 2D Dirac material #
##################################

#----------#
# PREAMBLE #
#----------#

# Call necessary mathematical and numerical libraries
from numpy import frompyfunc, linspace, exp, log, pi, zeros, arange
from mpmath import polylog
from scipy.integrate import quad
from matplotlib.ticker import AutoMinorLocator

# Call plotting library
import matplotlib.pyplot as plt
%matplotlib inline

SMALL_SIZE  = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)        # controls default text sizes
plt.rc('legend', fontsize=SMALL_SIZE)  # legend fontsize
plt.rc('xtick', labelsize=MEDIUM_SIZE) # fontsize of the x tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE) # fontsize of the y tick labels
plt.rc('axes', labelsize=BIGGER_SIZE)  # fontsize of the x and y labels

# Specific treatment for polylogarithm function
plog = frompyfunc(polylog, 2, 1)
# numpy.frompyfunc(func, nin, nout)
# Takes an arbitrary Python function and returns a NumPy ufunc.
# Parameters:
#            func : An arbitrary Python function.
#            nin  : (int) The number of input arguments.
#            nout : (int) The number of objects returned by func.
# Returns a NumPy universal function (ufunc) object.

#-----------#
# MAIN PART #
#-----------#

# Define kernel:

infinity = 100  # this value is sufficient for our function

## conduction band

def F0c(x):
    return exp(x)/(1+exp(x))
def F1c(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2c(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def F3c(x):
    return ( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x)) - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )

def G0c(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D-xx, infinity, args=(xx))
    return I
def G1c(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D-xx, infinity, args=(xx))
    return I
def G2c(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D-xx, infinity, args=(xx))
    return I

def L0c(Eta,DD):
    return F1c(Eta - DD) + Eta * F0c(Eta - DD) - DD**2 * G0c(Eta,DD)
def L1c(Eta,DD):
    return F2c(Eta - DD) + Eta * F1c(Eta - DD) - DD**2 * G1c(Eta,DD)
def L2c(Eta,DD):
    return F3c(Eta - DD) + Eta * F2c(Eta - DD) - DD**2 * G2c(Eta,DD)

## valence band

def F0v(x):
    return 1/(1+exp(x))
def F1v(x):
    return ( -x/(1+exp(x)) - log(1+exp(-x)) )
def F2v(x):
    return ( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )
def F3v(x):
    return ( -(x**3)/(1+exp(x)) - 3*(x**2)*log(1+exp(-x)) + 6*x*plog(2,-exp(-x)) + 6*plog(3,-exp(-x)) )


def G0v(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D-xx, args=(xx))
    return I
def G1v(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D-xx, args=(xx))
    return I
def G2v(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D-xx, args=(xx))
    return I

def L0v(Eta,DD):
    return -1 * ( F1v(Eta + DD) + Eta * F0v(Eta + DD) - DD**2 * G0v(Eta,DD) )
def L1v(Eta,DD):
    return -1 * ( F2v(Eta + DD) + Eta * F1v(Eta + DD) - DD**2 * G1v(Eta,DD) )
def L2v(Eta,DD):
    return -1 * ( F3v(Eta + DD) + Eta * F2v(Eta + DD) - DD**2 * G2v(Eta,DD) ) 

# TE quantities:

def Sc(Eta,DD):
    return -L1c(Eta,DD) / L0c(Eta,DD)
def Sv(Eta,DD):
    return -L1v(Eta,DD) / L0v(Eta,DD)
def sigmac(Eta,DD):
    return L0c(Eta,DD)
def sigmav(Eta,DD):
    return L0v(Eta,DD)
def kappac(Eta,DD):
    return L2c(Eta,DD) - ((L1c(Eta,DD))**2 / L0c(Eta,DD))
def kappav(Eta,DD):
    return L2v(Eta,DD) - ((L1v(Eta,DD))**2 / L0v(Eta,DD))

def sigma(Eta,DD):
    return sigmac(Eta,DD) + sigmav(Eta,DD)
def Seebeck(Eta,DD):
    return ((sigmac(Eta,DD) * Sc(Eta,DD) +
            sigmav(Eta,DD) * Sv(Eta,DD)) / sigma(Eta,DD))
def kappae(Eta,DD):
    return (kappac(Eta,DD) + kappav(Eta,DD) + 
            (sigmac(Eta,DD) * sigmav(Eta,DD))*
            (Sc(Eta,DD) - Sv(Eta,DD))**2 / sigma(Eta,DD))
def ZT(Seeb,Sig,KappaEl,KappaPh):
    return (Seeb**2 * Sig)/(KappaEl + KappaPh)

# Create sample points within [-10,10]:
npoint = 101
eta    = linspace(-10, 10, npoint) 

delta0 = 0.002
delta1 = 2.0
delta2 = 4.0
delta3 = 6.0
delta4 = 8.0

# Define the size of the entire plot
plt.figure(figsize=(6,4))

# Subplot (a): Seebeck's coefficient
ax1 = plt.subplot(2,2,1)
S0 = zeros(npoint,float)
S1 = zeros(npoint,float)
S2 = zeros(npoint,float)
S3 = zeros(npoint,float)
S4 = zeros(npoint,float)
for i in range(eta.size):
    S0[i] = Seebeck(eta[i],delta0)
    S1[i] = Seebeck(eta[i],delta1)
    S2[i] = Seebeck(eta[i],delta2)
    S3[i] = Seebeck(eta[i],delta3)
    S4[i] = Seebeck(eta[i],delta4)
plt.plot(eta, S0, color='red', linestyle='-',linewidth=1.5)
plt.plot(eta, S1, color='royalblue', linestyle='-',linewidth=1.5)
plt.plot(eta, S2, color='green', linestyle='-',linewidth=1.5)
plt.plot(eta, S3, color='brown', linestyle='-',linewidth=1.5)
plt.plot(eta, S4, color='teal', linestyle='-',linewidth=1.5)
plt.axis([-10,10,-8,8]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-10,11,5))
ax1.xaxis.set_minor_locator(AutoMinorLocator(5))
plt.yticks(range(-8,9,4))
ax1.yaxis.set_minor_locator(AutoMinorLocator(4))
ax1.yaxis.labelpad = -4
plt.ylabel('$S$ $(S_0)$')
plt.text(-16,7.2,'(a)')
plt.text(-1.5,-1,r'$0$',fontsize=10,color='red')
plt.text(-2.2,2.3,r'$2$',fontsize=10,color='royalblue')
plt.text(-3.5,3.5,r'$4$',fontsize=10,color='green')
plt.text(-4.4,4.5,r'$6$',fontsize=10,color='brown')
plt.text(-5.6,5.6,r'$8$',fontsize=10,color='teal')

# Subplot (b): Electrical conductivity
ax2 = plt.subplot(2,2,2)
Sigmae0 = zeros(npoint,float)
Sigmae1 = zeros(npoint,float)
Sigmae2 = zeros(npoint,float)
Sigmae3 = zeros(npoint,float)
Sigmae4 = zeros(npoint,float)
for i in range(eta.size):
    Sigmae0[i] = sigma(eta[i],delta0)
    Sigmae1[i] = sigma(eta[i],delta1)
    Sigmae2[i] = sigma(eta[i],delta2)
    Sigmae3[i] = sigma(eta[i],delta3)
    Sigmae4[i] = sigma(eta[i],delta4)
plt.plot(eta, Sigmae0, color='red', linestyle='-', linewidth=1.5)
plt.plot(eta, Sigmae1, color='royalblue', linestyle='-', linewidth=1.5)
plt.plot(eta, Sigmae2, color='green', linestyle='-', linewidth=1.5)
plt.plot(eta, Sigmae3, color='brown', linestyle='-', linewidth=1.5)
plt.plot(eta, Sigmae4, color='teal', linestyle='-', linewidth=1.5)
plt.axis([-10,10,0,6]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-10,11,5))
ax2.xaxis.set_minor_locator(AutoMinorLocator(5))
plt.yticks(arange(0,7,step=2))
ax2.yaxis.set_minor_locator(AutoMinorLocator(5))
plt.ylabel('$\sigma$ $(\sigma_0)$')
plt.text(-15,5.7,'(b)')
plt.text(-0.5,0.77,r'$0$',fontsize=10,color='red')
plt.text(2,0.59,r'$2$',fontsize=10,color='royalblue')
plt.text(3.7,0.6,r'$4$',fontsize=10,color='green')
plt.text(5.85,0.6,r'$6$',fontsize=10,color='brown')
plt.text(8,0.6,r'$8$',fontsize=10,color='teal')

# Subplot (c): Thermal conductivity (electron part)
ax3 = plt.subplot(2,2,3)
Kappae0 = zeros(npoint,float)
Kappae1 = zeros(npoint,float)
Kappae2 = zeros(npoint,float)
Kappae3 = zeros(npoint,float)
Kappae4 = zeros(npoint,float)
for i in range(eta.size):
    Kappae0[i] = kappae(eta[i],delta0)
    Kappae1[i] = kappae(eta[i],delta1)
    Kappae2[i] = kappae(eta[i],delta2)
    Kappae3[i] = kappae(eta[i],delta3)
    Kappae4[i] = kappae(eta[i],delta4)
plt.plot(eta, Kappae0, color='red', linestyle='-', linewidth=1.5)
plt.plot(eta, Kappae1, color='royalblue', linestyle='-', linewidth=1.5)
plt.plot(eta, Kappae2, color='green', linestyle='-', linewidth=1.5)
plt.plot(eta, Kappae3, color='brown', linestyle='-', linewidth=1.5)
plt.plot(eta, Kappae4, color='teal', linestyle='-', linewidth=1.5)
plt.axis([-10,10,0,16]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-10,11,5))
ax3.xaxis.set_minor_locator(AutoMinorLocator(5))
plt.yticks(range(0,17,4))
ax3.yaxis.set_minor_locator(AutoMinorLocator(4))
plt.xlabel('$\mu$ $(k_B T)$')
plt.ylabel('$\kappa_e$ $(\kappa_0)$')
plt.text(-16,15,'(c)')
plt.text(-0.5,8,r'$0$',fontsize=10,color='red')
plt.text(2,6,r'$2$',fontsize=10,color='royalblue')
plt.text(3.6,5,r'$4$',fontsize=10,color='green')
plt.text(5.5,4,r'$6$',fontsize=10,color='brown')
plt.text(7,3.2,r'$8$',fontsize=10,color='teal')
ax3.yaxis.labelpad = -2

# Subplot (d): Figure of merit
ax4 = plt.subplot(2,2,4)
rkapp = 1
plt.plot(eta, ZT(S0, Sigmae0, Kappae0, rkapp), color='red', linestyle='-', linewidth=1.5)
plt.plot(eta, ZT(S1, Sigmae1, Kappae1, rkapp), color='royalblue', linestyle='-', linewidth=1.5)
plt.plot(eta, ZT(S2, Sigmae2, Kappae2, rkapp), color='green', linestyle='-', linewidth=1.5)
plt.plot(eta, ZT(S3, Sigmae3, Kappae3, rkapp), color='brown', linestyle='-', linewidth=1.5)
plt.plot(eta, ZT(S4, Sigmae4, Kappae4, rkapp), color='teal', linestyle='-', linewidth=1.5)
plt.axis([-10,10,0,4]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-10,11,5))
ax4.xaxis.set_minor_locator(AutoMinorLocator(5))
plt.yticks(arange(0,4.1,step=1))
ax4.yaxis.set_minor_locator(AutoMinorLocator(4))
plt.text(4.5,3.5, r'$r_\kappa = 1$')
plt.text(-15,3.75,'(d)')
plt.xlabel('$\mu$ $(k_B T)$')
plt.ylabel('$ZT$')
plt.text(-0.5,0.2,r'$0$',fontsize=10,color='red')
plt.text(-1.2,1,r'$2$',fontsize=10,color='royalblue')
plt.text(-3,2.2,r'$4$',fontsize=10,color='green')
plt.text(-5,2.7,r'$6$',fontsize=10,color='brown')
plt.text(-7,2.8,r'$8$',fontsize=10,color='teal')

# Save figure
plt.tight_layout(pad=0.1, w_pad=0.5, h_pad=0.1)
plt.savefig('CRTfigTEvarygap.pdf')
```

```python
# Create sample points:
neta   = 1001
ndelta = 81
nkapp  = 11
eta    = linspace(-100,100,neta) 
delta  = linspace(0,80,ndelta)
rkapp  = linspace(0,10,nkapp)

# exception for delta = 0 and rkapp = 0
# due to numerical issue
delta[0] = 0.002
rkapp[0] = 0.002

# Initialize TE properties
S      = zeros((neta,ndelta))
Sigmae = zeros((neta,ndelta))
Kappae = zeros((neta,ndelta))
ZTcal  = zeros((neta,ndelta,nkapp))

# Calculate TE properties within the sample points
for i in range(eta.size):
    for j in range(delta.size):
        S[i,j]      = Seebeck(eta[i],delta[j])
        Sigmae[i,j] = sigma(eta[i],delta[j])
        Kappae[i,j] = kappae(eta[i],delta[j])
        for k in range(rkapp.size):
            ZTcal[i,j,k] = ZT(S[i,j],Sigmae[i,j],Kappae[i,j],rkapp[k])

# Finding maximum
ZTopt = zeros((ndelta,nkapp))
ZTeta = zeros(neta,float)
for j in range(delta.size):
    for k in range(rkapp.size):
        for i in range(eta.size):
            ZTeta[i] = ZTcal[i,j,k]
        ZTopt[j,k] = max(ZTeta)
```
