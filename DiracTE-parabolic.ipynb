# Comparing thermoelectrics of parabolic band with Dirac band

Given the energy dispersion for 2D parabolic band as,
\begin{equation}
E=\pm \left(\frac{\hbar^2|\textbf{k}|^2}{2m}+\Delta\right)
\end{equation}

For that band structure, the longitudinal velocity is given by,
\begin{equation}
v^2=\frac{v_g^2}{2}=\frac{E(k)}{m^*},\quad E(k)=\frac{\hbar^2k^2}{2m^*}.
\end{equation}

For conduction band, $E(k)=E-\Delta$ and for valence band, $E(k)=-E-\Delta$.



## 1/DOS relaxation time

For $\tau(E)=C[\mathcal{D}(E)]^{-1}$,
the transport distribution function can be expressed below,
\begin{align}
\mathcal{T}(E)&=v^2(E)\tau(E)\mathcal{D}(E) \\
&=Cv^2 \\
&=C\frac{E(k)}{m^*}
\end{align}

### Conduction band

The kernel function will become,
	\begin{align}
	\mathcal{L}_{i,c}&=\int\limits_{\Delta}^{\infty}dE\,\mathcal{T}(E)(E-\mu)^i\left(-\frac{\partial f_0}{\partial E}\right) \nonumber \\
	&=\frac{C}{m^*}\int\limits_{\Delta}^{\infty}dE\,(E-\Delta)(E-\mu)^i\left(-\frac{\partial f_0}{\partial E}\right).
	\end{align}
    
Substitute the Fermi-Dirac distribution function $f_0$ so we will get,
	\begin{align}
	\mathcal{L}_{i,c}=\frac{C}{m^*}\int\limits_{\Delta}^{\infty}dE\,(E-\Delta)(E-\mu)^i\frac{1}{k_BT}\frac{\exp{\frac{E-\mu}{k_BT}}}{(\exp{\frac{E-\mu}{k_BT}}+1)^2}.
	\end{align}
    
Define,
	\begin{align}\label{eq:DefinisiVarep&Eta}
	\varepsilon=E/(k_BT), \quad \eta=\mu/(k_BT), \quad \tilde{\Delta}=\Delta/(k_BT)
	\end{align}
so, $d\varepsilon=dE/(k_BT)$ and previous equation turns into,
	\begin{align}
	\mathcal{L}_{i,c}&=\frac{C}{m^*}(k_BT)^{i+1}
    \int\limits_{\tilde{\Delta}}^{\infty}d\varepsilon\,(\varepsilon-\tilde{\Delta})(\varepsilon-\eta)^i\frac{\exp{(\varepsilon-\eta)}}{(\exp{(\varepsilon-\mu)}+1)^2} \nonumber \\
	&=\frac{C}{m^*}(k_BT)^{i+1}\int_{\tilde{\Delta}-\eta}^\infty dx\,x^i(x+\eta-\tilde{\Delta})\frac{\exp(x)}{(\exp(x)+1)^2}, \\
    &=\frac{C}{m^*}(k_BT)^{i+1}\left(\mathcal{F}_{i+1,c}(\eta-\tilde{\Delta})+(\eta-\tilde{\Delta})\mathcal{F}_{i,c}(\eta-\tilde{\Delta})\right),
	\end{align}
with $\mathcal{F}_{i,c}(\eta)$ expressed as below,
	\begin{align}
\mathcal{F}_{i,c}(\eta)=\int_{-\eta}^\infty dx\,x^i\frac{\exp(x)}{(\exp(x)+1)^2}.
	\end{align}


### Valence band

Identical to the conduction band,
\begin{align}
	\mathcal{L}_{i,v}&=\frac{C}{m^*}\int\limits_{-\infty}^{-\Delta}dE\,(-E-\Delta)(E-\mu)^i\frac{1}{k_BT}\frac{\exp{\frac{E-\mu}{k_BT}}}{(\exp{\frac{E-\mu}{k_BT}}+1)^2}\\
    &=\frac{C}{m^*}(k_BT)^{i+1}\int\limits_{-\infty}^{-\tilde{\Delta}}d\varepsilon\,(-\varepsilon-\tilde{\Delta})(\varepsilon-\eta)^i\frac{\exp{(\varepsilon-\eta)}}{(\exp{(\varepsilon-\mu)}+1)^2} \nonumber \\
    &=\frac{C}{m^*}(k_BT)^{i+1}\left(\mathcal{F}_{i+1,v}(\eta+\tilde{\Delta})+(\eta+\tilde{\Delta})\mathcal{F}_{i,v}(\eta+\tilde{\Delta})\right),
\end{align}
with $\mathcal{F}_{i,v}(\eta)$ expressed as,
	\begin{align}
\mathcal{F}_{i,v}(\eta)=-\int_{-\infty}^{-\eta} dx\,x^i\frac{\exp(x)}{(\exp(x)+1)^2}.
	\end{align}

### The thermoelectric coefficient

\begin{align}
    \sigma_\alpha&=q^2\mathcal{L}_{0,\alpha} \\
        &=q^2\frac{C}{m^*}(k_BT)\left(\mathcal{F}_{1,\alpha}(\eta+s_\alpha\tilde\Delta)+(\eta+s_\alpha\tilde\Delta)\mathcal{F}_{0,\alpha}(\eta+s_\alpha\Delta)\right) \\
        &\equiv\sigma_0\left(\mathcal{F}_{1,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{0,\alpha}(\eta+s_\alpha\Delta)\right),\\
\\
    S_\alpha&=-\frac{1}{qT}\frac{\mathcal{L}_{1,\alpha}}{\mathcal{L}_{0,\alpha}} \\
       &=-\frac{k_B}{q}\frac{\left(\mathcal{F}_{2,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{1,\alpha}(\eta+s_\alpha\Delta)\right)}{\left(\mathcal{F}_{1,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{0,\alpha}(\eta+s_\alpha\Delta)\right)} \\
       &\equiv-S_0\frac{\left(\mathcal{F}_{2,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{1,\alpha}(\eta+s_\alpha\Delta)\right)}{\left(\mathcal{F}_{1,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{0,\alpha}(\eta+s_\alpha\Delta)\right)},\\
\\
    \kappa_{e,\alpha} &= \frac{1}{T} \left(\mathcal{L}_{2,\alpha} -
\frac{(\mathcal{L}_{1,\alpha})^2}{\mathcal{L}_{0,\alpha}}\right)\\
        &=\frac{Ck_B^3T^2}{m^*}
            \left(\left(\mathcal{F}_{3,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{2,\alpha}(\eta+s_\alpha\Delta)\right)-\frac{\left(\mathcal{F}_{2,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{1,\alpha}(\eta+s_\alpha\Delta)\right)^2}{\left(\mathcal{F}_{1,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{0,\alpha}(\eta+s_\alpha\Delta)\right)}
            \right) \\
        &=\kappa_0
            \left(\left(\mathcal{F}_{3,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{2,\alpha}(\eta+s_\alpha\Delta)\right)-\frac{\left(\mathcal{F}_{2,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{1,\alpha}(\eta+s_\alpha\Delta)\right)^2}{\left(\mathcal{F}_{1,\alpha}(\eta+s_\alpha\Delta)+(\eta+s_\alpha\Delta)\mathcal{F}_{0,\alpha}(\eta+s_\alpha\Delta)\right)}
            \right),
\end{align}

with,
$\sigma_0=\frac{Cq^2k_BT}{m^*}$, $S_0=\frac{k_B}{q}$, $\kappa_0=\frac{Ck_B^3T^2}{m^*}$, and $s_\alpha$ is $1$ when $\alpha=v$ (valence band) or $-1$ when $\alpha=c$ (conduction band).

## Constant relaxation time

For constant relaxatio time $\tau(E)=\tau_0$ and density of states given by,
\begin{align}
    d(E)&=g\sum_{\textbf{k}}\delta(E_\textbf{k}-E) \\
    \mathcal{D}(E)&=\frac{g}{(2\pi)^2}\int d^2\textbf{k}\delta(E_\textbf{k}-E), \quad E_\textbf{k}=\frac{\hbar^2k^2}{2m^*}+\Delta^2\\
    &=\frac{g}{(2\pi)^2}2\pi\int k\,dk\,\delta(E_\textbf{k}E),\quad kdk=\frac{m^*}{\hbar^2}dE_\textbf{k} \\
    &=\frac{g}{2\pi}\frac{m^*}{\hbar^2}\int_{-\infty}^{\infty}dE_\textbf{k}\delta(E_\textbf{k}-E) \\
    &=\frac{gm^*}{2\pi\hbar^2}\Theta(|E|-|\Delta|),
\end{align}
where $\Theta(x)$ is a step function, $\Theta(x)=1$ if $x>0$ and $\Theta(x)=0$ otherwise, and $g$ is the degeneracy factor, which is 2 to account for the spin degree of freedom for fermion.
Then, the TDF can be written as,
\begin{align}
\mathcal{T}(E)&=\tau_0\left(\frac{E(k)}{m^*}\right)\left[\frac{gm^*}{2\pi\hbar^2}\Theta\left(|E|-|\Delta|\right)\right] \\
&=\tau_0\left(\frac{gE(k)}{2\pi\hbar^2}\Theta\left(|E|-|\Delta|\right)\right)
\end{align}

### Conduction band

\begin{align}
	\mathcal{L}_{i,c}&=\int\limits_{\Delta}^{\infty}dE\,\mathcal{T}(E)(E-\mu)^i\left(-\frac{\partial f_0}{\partial E}\right) \nonumber \\
	&=\frac{\tau_0g}{2\pi\hbar^2}\int\limits_{\Delta}^{\infty}dE\,(E-\Delta)(E-\mu)^i\left(-\frac{\partial f_0}{\partial E}\right) \nonumber \\
    &=\frac{\tau_0g}{2\pi\hbar^2}(k_BT)^{i+1}\int\limits_{\tilde{\Delta}}^{\infty}d\varepsilon\,(\varepsilon-\tilde\Delta)(\varepsilon-\eta)^i\frac{\exp{(\varepsilon-\eta)}}{(\exp{(\varepsilon-\mu)}+1)^2} \nonumber \\
	&=\frac{\tau_0g}{2\pi\hbar^2}(k_BT)^{i+1}\int_{\tilde{\Delta}-\eta}^\infty dx\,x^i(x+\eta-\tilde\Delta)\frac{\exp(x)}{(\exp(x)+1)^2}, \\
    &=\frac{\tau_0g}{2\pi\hbar^2}(k_BT)^{i+1}\left(\mathcal{F}_{i+1,c}(\eta-\tilde{\Delta})+(\eta-\tilde\Delta)\mathcal{F}_{i,c}(\eta-\tilde{\Delta})\right),
\end{align}

### Valence band

Identical to the conduction band,
\begin{align}
	\mathcal{L}_{i,v}&=\frac{\tau_0g}{2\pi\hbar^2}\int\limits_{-\infty}^{-\Delta}dE\,(-E-\Delta)(E-\mu)^i\frac{1}{k_BT}\frac{\exp{\frac{E-\mu}{k_BT}}}{(\exp{\frac{E-\mu}{k_BT}}+1)^2}\\
    &=\frac{\tau_0g}{2\pi\hbar^2}(k_BT)^{i+1}\int\limits_{-\infty}^{-\tilde{\Delta}}d\varepsilon\,(-\varepsilon-\tilde\Delta)(\varepsilon-\eta)^i\frac{\exp{(\varepsilon-\eta)}}{(\exp{(\varepsilon-\mu)}+1)^2} \nonumber \\
    &=\frac{\tau_0g}{2\pi\hbar^2}(k_BT)^{i+1}\left(\mathcal{F}_{i+1,v}(\eta+\tilde{\Delta})+(\eta+\tilde\Delta)\mathcal{F}_{i,v}(\eta+\tilde{\Delta})\right),
\end{align}
with $\mathcal{F}_{i,v}(\eta)$:
	\begin{align}
\mathcal{F}_{i,v}(\eta)=-\int_{-\infty}^{-\eta} dx\,x^i\frac{\exp(x)}{(\exp(x)+1)^2}.
	\end{align}


### The thermoelectric coefficients
The expression to the TE quantities in terms of the $\mathcal{F}$ integral is still the same with using 1/DOS relaxation time approximation, difference is in its unit, $\sigma_0=\frac{\tau_0g}{2\pi\hbar^2}q^2k_BT$, and $\kappa_0=\frac{\tau_0g}{2\pi\hbar^2}k_B^3T^2$:

# Optimal gap and chemical potential for parabolic bands

```python
####################################
# TE of 2D Parabolic band material #
####################################

#----------#
# PREAMBLE #
#----------#

# Call necessary mathematical and numerical libraries
from numpy import frompyfunc, linspace, exp, log, pi, zeros, arange, argmax
from mpmath import polylog
from scipy.integrate import quad
from matplotlib.ticker import AutoMinorLocator

# Call plotting library
import matplotlib.pyplot as plt
%matplotlib inline

SMALL_SIZE  = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)        # controls default text sizes
plt.rc('legend', fontsize=SMALL_SIZE)  # legend fontsize
plt.rc('xtick', labelsize=MEDIUM_SIZE) # fontsize of the x tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE) # fontsize of the y tick labels
plt.rc('axes', labelsize=BIGGER_SIZE)  # fontsize of the x and y labels

# Specific treatment for polylogarithm function
plog = frompyfunc(polylog, 2, 1)
# numpy.frompyfunc(func, nin, nout)
# Takes an arbitrary Python function and returns a NumPy ufunc.
# Parameters:
#            func : An arbitrary Python function.
#            nin  : (int) The number of input arguments.
#            nout : (int) The number of objects returned by func.
# Returns a NumPy universal function (ufunc) object.

#-----------#
# MAIN PART #
#-----------#

# Define kernel:

## conduction band
small = exp(-15)
def F0c(x):
    return 2*exp(x)/(1+exp(x))
def F1c(x):
    return 2*( x/(1+exp(x)) + log(1+exp(-x)) )
def F2c(x):
    return 2*( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def F3c(x):
    return 2*( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x)) - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )

## valence band

def F0v(x):
    return -2/(1+exp(x))
def F1v(x):
    return 2*( x/(1+exp(x)) + log(1+exp(-x)) )
def F2v(x):
    return -2*( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )
def F3v(x):
    return 2*( (x**3)/(1+exp(x)) + 3*(x**2)*log(1+exp(-x)) - 6*x*plog(2,-exp(-x)) - 6*plog(3,-exp(-x)) )

# TE quantities:

def Sc(Eta,DD):
    return -(F2c(Eta-DD)+(Eta-DD)*F1c(Eta-DD)) / (F1c(Eta-DD)+(Eta-DD)*F0c(Eta-DD)+small)
def Sv(Eta,DD):
    return -(F2v(Eta+DD)+(Eta+DD)*F1v(Eta+DD)) / (F1v(Eta+DD)+(Eta+DD)*F0v(Eta+DD)+small)
def sigmac(Eta,DD):
    return F1c(Eta-DD)+(Eta-DD)*F0c(Eta-DD)
def sigmav(Eta,DD):
    return F1v(Eta+DD)+(Eta+DD)*F0v(Eta+DD)
def kappac(Eta,DD):
    return (F3c(Eta-DD)+(Eta-DD)*F2c(Eta-DD)) - ((F2c(Eta-DD)+(Eta-DD)*F1c(Eta-DD))**2/(F1c(Eta-DD)+(Eta-DD)*F0c(Eta-DD)+small) )
def kappav(Eta,DD):
    return (F3v(Eta+DD)+(Eta+DD)*F2v(Eta+DD)) - ((F2v(Eta+DD)+(Eta+DD)*F1v(Eta+DD))**2/(F1v(Eta+DD)+(Eta+DD)*F0v(Eta+DD)+small) )

def sigma(Eta,DD):
    return sigmac(Eta,DD) + sigmav(Eta,DD)
def Seebeck(Eta,DD):
    return ((sigmac(Eta,DD) * Sc(Eta,DD) +
            sigmav(Eta,DD) * Sv(Eta,DD)) / (sigmac(Eta,DD) + sigmav(Eta,DD)+small))
def kappae(Eta,DD):
    return ((kappac(Eta,DD) + kappav(Eta,DD)) + 
            (sigmac(Eta,DD) * sigmav(Eta,DD))*
            (Sc(Eta,DD) - Sv(Eta,DD))**2 / (sigmac(Eta,DD) + sigmav(Eta,DD)+small))
def ZT(Seeb,Sig,KappaEl,KappaPh):
    return (Seeb**2 * Sig)/(KappaEl + KappaPh)

# Create sample points within [-10,10]:
npoint = 101
eta    = linspace(-10, 10, npoint) 

delta0 = 0.002
delta1 = 2.0
delta2 = 4.0
delta3 = 6.0
delta4 = 8.0

# Define the size of the entire plot
plt.figure(figsize=(6,4))

# Subplot (a): Seebeck's coefficient
ax1 = plt.subplot(2,2,1)
S0 = zeros(npoint,float)
S1 = zeros(npoint,float)
S2 = zeros(npoint,float)
S3 = zeros(npoint,float)
S4 = zeros(npoint,float)
for i in range(eta.size):
    S0[i] = Seebeck(eta[i],delta0)
    S1[i] = Seebeck(eta[i],delta1)
    S2[i] = Seebeck(eta[i],delta2)
    S3[i] = Seebeck(eta[i],delta3)
    S4[i] = Seebeck(eta[i],delta4)
plt.plot(eta, S0, color='red', linestyle='-',linewidth=1.5)
plt.plot(eta, S1, color='royalblue', linestyle='-',linewidth=1.5)
plt.plot(eta, S2, color='green', linestyle='-',linewidth=1.5)
plt.plot(eta, S3, color='brown', linestyle='-',linewidth=1.5)
plt.plot(eta, S4, color='teal', linestyle='-',linewidth=1.5)
plt.axis([-10,10,-8,8]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-10,11,5))
ax1.xaxis.set_minor_locator(AutoMinorLocator(5))
plt.yticks(range(-8,9,4))
ax1.yaxis.set_minor_locator(AutoMinorLocator(4))
ax1.yaxis.labelpad = -4
plt.ylabel('$S$ $(S_0)$')
plt.text(-16,7.2,'(a)')
plt.text(-1.5,-1,r'$0$',fontsize=10,color='red')
plt.text(-2.2,2.3,r'$2$',fontsize=10,color='royalblue')
plt.text(-3.5,3.5,r'$4$',fontsize=10,color='green')
plt.text(-4.4,4.5,r'$6$',fontsize=10,color='brown')
plt.text(-5.6,5.6,r'$8$',fontsize=10,color='teal')

# Subplot (b): Electrical conductivity
ax2 = plt.subplot(2,2,2)
Sigmae0 = zeros(npoint,float)
Sigmae1 = zeros(npoint,float)
Sigmae2 = zeros(npoint,float)
Sigmae3 = zeros(npoint,float)
Sigmae4 = zeros(npoint,float)
for i in range(eta.size):
    Sigmae0[i] = sigma(eta[i],delta0)
    Sigmae1[i] = sigma(eta[i],delta1)
    Sigmae2[i] = sigma(eta[i],delta2)
    Sigmae3[i] = sigma(eta[i],delta3)
    Sigmae4[i] = sigma(eta[i],delta4)
plt.plot(eta, Sigmae0, color='red', linestyle='-', linewidth=1.5)
plt.plot(eta, Sigmae1, color='royalblue', linestyle='-', linewidth=1.5)
plt.plot(eta, Sigmae2, color='green', linestyle='-', linewidth=1.5)
plt.plot(eta, Sigmae3, color='brown', linestyle='-', linewidth=1.5)
plt.plot(eta, Sigmae4, color='teal', linestyle='-', linewidth=1.5)
plt.axis([-10,10,0,6]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-10,11,5))
ax2.xaxis.set_minor_locator(AutoMinorLocator(5))
plt.yticks(arange(0,7,step=2))
ax2.yaxis.set_minor_locator(AutoMinorLocator(5))
plt.ylabel('$\sigma$ $(\sigma_0)$')
plt.text(-15,5.7,'(b)')
plt.text(-0.5,0.77,r'$0$',fontsize=10,color='red')
plt.text(2.4,0.59,r'$2$',fontsize=10,color='royalblue')
plt.text(4.1,0.6,r'$4$',fontsize=10,color='green')
plt.text(6.25,0.6,r'$6$',fontsize=10,color='brown')
plt.text(8.3,0.6,r'$8$',fontsize=10,color='teal')

# Subplot (c): Thermal conductivity (electron part)
ax3 = plt.subplot(2,2,3)
Kappae0 = zeros(npoint,float)
Kappae1 = zeros(npoint,float)
Kappae2 = zeros(npoint,float)
Kappae3 = zeros(npoint,float)
Kappae4 = zeros(npoint,float)
for i in range(eta.size):
    Kappae0[i] = kappae(eta[i],delta0)
    Kappae1[i] = kappae(eta[i],delta1)
    Kappae2[i] = kappae(eta[i],delta2)
    Kappae3[i] = kappae(eta[i],delta3)
    Kappae4[i] = kappae(eta[i],delta4)
plt.plot(eta, Kappae0, color='red', linestyle='-', linewidth=1.5)
plt.plot(eta, Kappae1, color='royalblue', linestyle='-', linewidth=1.5)
plt.plot(eta, Kappae2, color='green', linestyle='-', linewidth=1.5)
plt.plot(eta, Kappae3, color='brown', linestyle='-', linewidth=1.5)
plt.plot(eta, Kappae4, color='teal', linestyle='-', linewidth=1.5)
plt.axis([-10,10,0,16]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-10,11,5))
ax3.xaxis.set_minor_locator(AutoMinorLocator(5))
plt.yticks(range(0,17,4))
ax3.yaxis.set_minor_locator(AutoMinorLocator(4))
plt.xlabel('$\mu$ $(k_B T)$')
plt.ylabel('$\kappa_e$ $(\kappa_0)$')
plt.text(-16,15,'(c)')
plt.text(-0.5,8,r'$0$',fontsize=10,color='red')
plt.text(2,4.5,r'$2$',fontsize=10,color='royalblue')
plt.text(3.6,3.5,r'$4$',fontsize=10,color='green')
plt.text(5.5,2.5,r'$6$',fontsize=10,color='brown')
plt.text(7,1.7,r'$8$',fontsize=10,color='teal')
ax3.yaxis.labelpad = -2

# Subplot (d): Figure of merit
ax4 = plt.subplot(2,2,4)
rkapp = 1
plt.plot(eta, ZT(S0, Sigmae0, Kappae0, rkapp), color='red', linestyle='-', linewidth=1.5)
plt.plot(eta, ZT(S1, Sigmae1, Kappae1, rkapp), color='royalblue', linestyle='-', linewidth=1.5)
plt.plot(eta, ZT(S2, Sigmae2, Kappae2, rkapp), color='green', linestyle='-', linewidth=1.5)
plt.plot(eta, ZT(S3, Sigmae3, Kappae3, rkapp), color='brown', linestyle='-', linewidth=1.5)
plt.plot(eta, ZT(S4, Sigmae4, Kappae4, rkapp), color='teal', linestyle='-', linewidth=1.5)
plt.axis([-10,10,0,4]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-10,11,5))
ax4.xaxis.set_minor_locator(AutoMinorLocator(5))
plt.yticks(arange(0,4.1,step=1))
ax4.yaxis.set_minor_locator(AutoMinorLocator(4))
plt.text(4.5,3.5, r'$r_\kappa = 1$')
plt.text(-15,3.75,'(d)')
plt.xlabel('$\mu$ $(k_B T)$')
plt.ylabel('$ZT$')
plt.text(-0.5,0.2,r'$0$',fontsize=10,color='red')
plt.text(-1.2,1,r'$2$',fontsize=10,color='royalblue')
plt.text(-3,2,r'$4$',fontsize=10,color='green')
plt.text(-5,2.2,r'$6$',fontsize=10,color='brown')
plt.text(-7,2.2,r'$8$',fontsize=10,color='teal')

# Save figure
plt.tight_layout(pad=0.1, w_pad=0.5, h_pad=0.1)
plt.savefig('parabolFigTEvarygap.pdf')
```

```python
# Create sample points:
neta   = 251
ndelta = 21
nkapp  = 6
eta    = linspace(0,25,neta) 
delta  = linspace(0,20,ndelta)
rkapp  = linspace(0,5,nkapp)

# exception for delta = 0 and rkapp = 0
# due to numerical issue
#delta[0] = 0.002
#rkapp[0] = 0.002

# Initialize TE properties
S      = zeros((neta,ndelta))
Sigmae = zeros((neta,ndelta))
Kappae = zeros((neta,ndelta))
ZTcal  = zeros((neta,ndelta,nkapp))

# Calculate TE properties within the sample points
for i in range(eta.size):
    for j in range(delta.size):
        S[i,j]      = Seebeck(eta[i],delta[j])
        Sigmae[i,j] = sigma(eta[i],delta[j])
        Kappae[i,j] = kappae(eta[i],delta[j])
        for k in range(rkapp.size):
            ZTcal[i,j,k] = ZT(S[i,j],Sigmae[i,j],Kappae[i,j],rkapp[k])

# Finding maximum
ZTopt = zeros((ndelta,nkapp))
etaopt = zeros((ndelta,nkapp))
ZTeta = zeros(neta,float)
for j in range(delta.size):
    for k in range(rkapp.size):
        for i in range(eta.size):
            ZTeta[i] = ZTcal[i,j,k]
        ZTopt[j,k] = max(ZTeta)
        etaopt[j,k]=abs(eta[argmax(ZTeta)])
```
