# Boltzmann's transport theory + relaxation time approximation

Suppose a material have energy dispersion $E(k)$. Within Boltzmann's transport theory and relaxation time  approximation (RTA), thermoelectric (TE) properties can be calculated by starting from the TE kernel (integral): 
\begin{align}
\mathcal{L}_i = \int_{-\infty}^\infty \mathcal{T} (E) (E-\mu)^i
\left(-\frac{\partial f}{\partial E}\right) dE, \quad (i = 0, 1, 2)
\end{align}
where $\mu$ is the Fermi energy, $f(E)$ is the Fermi-Dirac distribution, and $\mathcal{T}(E)$ is the transport distribution function (TDF).

The explicit form of TDF is:
\begin{align}
\mathcal{T}(E)=v_x^2(E)\tau(E)\mathcal{D}(E),
\end{align}
where $v_x$ is the longitudinal velocity, $\tau$ is the relaxation time, and $\mathcal {D}(E)$ is the density of states (DOS).  The longitudinal velocity $v_x$ depends on material's dimension $d$ by:
\begin{align}
v_x^2 = \frac{v_g^2}{d},
\end{align}
where $v_g$ is the group velocity.

We assume that the relaxation time from Fermi's golden rule is inversely proportional to the DOS:
\begin{align}
\tau(E) = C \left(\mathcal{D}(E)\right)^{-1},
\end{align}
where $C$ is a scattering coefficient in units of $\mathrm{W}^{-1}\mathrm{m}^{-3}$.

Using the TE kernel, one can find all TE quantities that originate from the electron transport.  The electrical conductivity $\sigma$, Seebeck's coefficient $S$, and electron thermal conductivity $\kappa_e$ are respectively given by:
\begin{align}
\sigma &= q^2\mathcal{L}_0,\\
S &= \frac{1}{qT} \frac{\mathcal{L}_1}{\mathcal{L}_0},\\
\kappa_e &= \frac{1}{T}\left(\mathcal{L}_2-
\frac{\left(\mathcal{L}_1\right)^2}{\mathcal{L}_0}\right),
\end{align}
where $T$ is the absolute temperature.  The figure of merit $ZT$ is then given by:
\begin{align}
ZT &= \frac{S^2 \sigma}{\kappa_e + \kappa_{\mathrm{ph}}} T,
\end{align}
where $\kappa_{\mathrm{ph}}$ is the lattice (phonon) thermal conductivity. 

Later when simplifying some equations, we will use the reduced variables for the energy dispersion:
\begin{equation}
\varepsilon = \displaystyle\frac{E}{k_B T},
\end{equation}
and for the chemical potential:
\begin{equation}
\eta = \displaystyle\frac{\mu}{k_B T},
\end{equation}
where $k_B$ is the Boltzmann constant.

## Gapless 2D Dirac material

We consider a gapless 2D Dirac material ($d = 2$) with energy dispersion $E = \pm \hbar v_F|\mathbf{k}|$, where $v_F$ is the Fermi velocity in graphene.  In this case, the TDF is independent of $E$ because the group velocity $v_g = v_F$ is constant, so that
\begin{align}
\mathcal{T} &= \frac{v_F^2 C}{d}\\
            &= \frac{1}{2} v_F^2 C.
\end{align}

### Gapless one band model

In the one band model, the integral $\mathcal{L}_i$ is calculated within $[0,\infty]$ considering only a gapless conduction band.  Therefore, we rewrite $\mathcal{L}_i$ as
\begin{align}
\mathcal{L}_{i,c} = \int_{0}^\infty \mathcal{T} (E) (E-\mu)^i
\left(-\frac{\partial f}{\partial E}\right) dE, \quad (i = 0, 1, 2)
\end{align}
where the additional subscript $c$ labels the conduction band.

We can express the kernel more explicitly, 
\begin{align}
\mathcal{L}_{i,c} &= \mathcal{T} \int_0^\infty (E-\mu)^i \frac{1}{k_B T}
                        \frac{e^{(E-\mu)/(k_B T)}}{\left(e^{(E-\mu)/(k_B T)} + 1\right)^2} dE \\
                    &= \frac{1}{2}v_F^2 C (k_B T)^i \int_0^\infty (\varepsilon-\eta)^i
                        \frac{e^{(\varepsilon-\eta)}}{\left(e^{(\varepsilon-\eta)} +1
                        \right)^2} d\varepsilon \\
                    &= \frac{1}{2}v_F^2 C (k_B T)^i \mathcal{F}_{i,c}(\eta),
\end{align}
with
\begin{align}
\mathcal{F}_{i,c} (\eta) = \int_{-\eta}^\infty \frac{x^i e^{x}}{\left(e^{x} +1 \right)^2} dx, 
\end{align}
due to the substitution of $x = \varepsilon - \eta$.

For $i = 0,1,2$, we have:
\begin{align}
\mathcal{F}_{0,c} (\eta) &= \frac{e^\eta}{e^\eta + 1},\\
\mathcal{F}_{1,c} (\eta) &= \frac{\eta}{e^\eta + 1} + \ln(1 + e^{-\eta}),\\
\mathcal{F}_{2,c} (\eta) &= \frac{\pi^2}{3} - \frac{\eta^2}{e^\eta + 1} - 2\eta \ln(1 + e^{-\eta}) + 2 \mathrm{Li}_2(-e^{-\eta}),
\end{align}
where $\mathrm{Li}_k(z) = \displaystyle\sum_{n=1}^\infty \frac{z^n}{n^k}$ is the polylogarithmic function.

Now we can obtain $\sigma_c$, $S_c$, and $\kappa_{e,c}$ as follows:
\begin{align}
\sigma_c &= q^2 \mathcal{L}_{0,c}\\
            &= \frac{1}{2} e^2 v_F^2 C\mathcal{F}_{0,c} (\eta),\\
S_c &= \frac{1}{qT} \frac{\mathcal{L}_{1,c}}{\mathcal{L}_{0,c}}\\
            &= -\frac{k_B}{e} \frac{\mathcal{F}_{1,c}(\eta)}{\mathcal{F}_{0,c}(\eta)},\\
\kappa_{e,c} &= \frac{1}{T} \left(\mathcal{L}_{2,c} -
\frac{(\mathcal{L}_{1,c})^2}{\mathcal{L}_{0,c}}\right)\\
            &= \frac{1}{2} v_F^2 C k_B^2 T \left(\mathcal{F}_{2,c} (\eta) -
                \frac{(\mathcal{F}_{1,c}(\eta))^2}{\mathcal{F}_{0,c}(\eta)} \right)\\
            &= \kappa_0 \left(\mathcal{F}_{2,c} (\eta) -
                \frac{(\mathcal{F}_{1,c}(\eta))^2}{\mathcal{F}_{0,c}(\eta)} \right),
\end{align}
where $e$ is the unit electron charge and $\kappa_0 = \displaystyle\frac{1}{2}v_F^2 C k_B^2 T$.  Here $\kappa_0$ has the same units as $\kappa_{e,c}$.

We can also express $\kappa_{\mathrm{ph}}$ in terms of $\kappa_0$, i.e.,
\begin{align}
\kappa_{\mathrm{ph}} = r_\kappa \kappa_0.
\end{align}
We expect that the parameter $r_\kappa$ depends on the material and may be engineered to maximize the $ZT$.  For the gapless 2D Dirac material within the one band model, we can write:
\begin{align}
ZT(\eta) &= \frac{S_c^2 \sigma_c}{\kappa_{e,c} + \kappa_{\mathrm{ph}}} T\\
&= \frac{(\mathcal{F}_{1,c}(\eta))^2}{\mathcal{F}_{0,c} (\eta)
\mathcal{F}_{2,c}(\eta) - (\mathcal{F}_{1,c} (\eta))^2 + r_\kappa \mathcal{F}_{0,c}}
\end{align}

### Code: gapless one band

```python
###################################################
# ZT of GAPLESS 2D Dirac material: ONE BAND model #
###################################################

#----------#
# PREAMBLE #
#----------#

# Libraries below will be used throughout the notebook.
# Therefore, we must execute this cell at least once.

# Call necessary mathematical and numerical libraries
from numpy import frompyfunc, linspace, exp, log, pi, zeros
from mpmath import polylog
from scipy.integrate import quad

# Call plotting library
import matplotlib.pyplot as plt
%matplotlib inline

# Specific treatment for polylogarithm function
plog = frompyfunc(polylog, 2, 1)
# numpy.frompyfunc(func, nin, nout)
# Takes an arbitrary Python function and returns a NumPy ufunc.
# Parameters:
#            func : An arbitrary Python function.
#            nin  : (int) The number of input arguments.
#            nout : (int) The number of objects returned by func.
# Returns a NumPy universal function (ufunc) object.

#-----------#
# MAIN PART #
#-----------#

# Define kernel:
def F0(x):
    return exp(x)/(1+exp(x))
def F1(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
# ZT function:
def ZT(Seeb,Sig,KappaE,KappaP):
    return (Seeb**2 * Sig)/(KappaE + KappaP)

# Create sample points within [-10,10]:
npoint  = 101
eta     = linspace(-10, 10, npoint) 

# TE quantities:
S_c     = -F1(eta) / F0(eta)
Sigma_c =  F0(eta)
kappa_c =  F2(eta) - ((F1(eta))**2 / F0(eta))

#----------#
# PLOTTING #
#----------#

# Example line formats are:
# 'b-'  = blue solid line
# 'r--' = red dashed line
# 'g:   = green dotted line
# if black is needed, use 'k' option (as in cmyk color)
plt.figure(figsize=(4,3))
plt.plot(eta, ZT(S_c, Sigma_c, kappa_c, 0), 'b-',
         eta, ZT(S_c, Sigma_c, kappa_c, 1), 'r--',
         eta, ZT(S_c, Sigma_c, kappa_c, 2), 'g:')
plt.axis([-4,4,0,2]) # the format is [xmin,xmax,ymin,ymax]
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-2.1,1.6, r'$r_\kappa = 0$', color='b')
plt.text(-2.1,1.1, r'$r_\kappa = 1$', color='r')
plt.text(-2.1,0.6, r'$r_\kappa = 2$', color='g')
# in plt.xlabel or ylabel, the option r is not necessary
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$ZT$')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()

# Save figure in PDF file
plt.savefig('ZTgapless1band.pdf')
```

### Gapless two band model

In the two band model, we have an additional term related to a valence band.  For the gapless 2D Dirac material, the TE kernel for the valence band is calculated within energy of $[-\infty,0]$: 
\begin{align}
\mathcal{L}_{i,v} = \int_{-\infty}^0 \mathcal{T} (E) (E-\mu)^i
\left(-\frac{\partial f}{\partial E}\right) dE, \quad (i = 0, 1, 2)
\end{align}
where the  subscript $v$ labels the conduction band.

Similar with the kernel for the conduction band, we can write the kernel for the valence band as follows: 
\begin{align}
\mathcal{L}_{i,v} = \frac{1}{2}v_F^2 C (k_B T)^i \mathcal{F}_{i,v}(\eta),
\end{align}
with
\begin{align}
\mathcal{F}_{i,v} (\eta) = \int_{-\infty}^{-\eta} \frac{x^i e^{x}}{\left(e^{x} +1 \right)^2} dx. 
\end{align}

For $i = 0,1,2$, we have:
\begin{align}
\mathcal{F}_{0,v} (\eta) &= \frac{1}{e^\eta + 1},\\
\mathcal{F}_{1,v} (\eta) &= - \frac{\eta}{e^\eta + 1} - \ln(1 + e^{-\eta}),\\
\mathcal{F}_{2,v} (\eta) &= \frac{\eta^2}{e^\eta + 1} + 2x \ln(1 + e^{-\eta}) - 2 \mathrm{Li}_2(-e^{-\eta}).                           
\end{align}
Note that the integrals $\mathcal{F}_{i,c}$ and $\mathcal{F}_{i,v}$ are connected via electron-hole symmetry of the system:
\begin{align}
\mathcal{F}_{0,c} (\eta) &= \mathcal{F}_{0,v} (-\eta)\\
\mathcal{F}_{1,c} (\eta) &= -\mathcal{F}_{1,v} (\eta)\\
\mathcal{F}_{2,c} (\eta) &= \mathcal{F}_{2,v} (-\eta)
\end{align}

The contribution of the valence band to the TE quantities, i.e., $\sigma_v$, $S_v$, and $\kappa_{e,v}$, can be expressed similarly with that of the conduction band:
\begin{align}
\sigma_v &= \frac{1}{2} e^2 v_F^2 C\mathcal{F}_{0,v} (\eta),\\
S_v &= - \frac{k_B}{e} \frac{\mathcal{F}_{1,v}(\eta)}{\mathcal{F}_{0,v}(\eta)},\\
\kappa_{e,v} &= \kappa_0 \left(\mathcal{F}_{2,v} (\eta) -
                \frac{(\mathcal{F}_{1,v}(\eta))^2}{\mathcal{F}_{0,v}(\eta)} \right)
\end{align}

In the two band model, we include both valence and conduction bands for each TE quantity.  The electrical conductivity $\sigma$, Seebeck's coefficient $S$, and electron thermal conductivity $\kappa_e$ can be calculated by the following formulas:
\begin{align}
\sigma &= \sigma_c + \sigma_v\\
S &= \frac{\sigma_c S_c + \sigma_v S_v}{\sigma_c + \sigma_v}\\
\kappa_e &= \frac{\sigma_c \sigma_v}{\sigma_c + \sigma_v}(S_c - S_v)^2
            + (\kappa_{e,c} + \kappa_{e,v})
\end{align}
The figure of merit $ZT(\eta)$ is then calculated in the same way as before.  However, due to the additional term in the two band model and also due to the electron-hole symmetry of the TE kernel, we find that Seebeck's coefficient $S$ becomes <b>zero</b>, whereas the electrical conductivity $\sigma$ and electron thermal conductivity $\kappa_e$ have nonzero (finite) values that are constant for whole $\eta$.  As a result,
\begin{align}
ZT(\eta) = 0
\end{align}
for the gapless 2D material in the two band model.

### Code: gapless two band

```python
###################################################
# ZT of GAPLESS 2D Dirac material: TWO BAND model #
###################################################

#----------#
# PREAMBLE #
#----------#

# We use the same preamble as in the one band model.
# Make sure the one band part has been executed! 
# (CTRL + ENTER) in that cell!

#-----------#
# MAIN PART #
#-----------#

# Define kernel:
## conduction band
def F0c(x):
    return exp(x)/(1+exp(x))
def F1c(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2c(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
## valence band
def F0v(x):
    return 1/(1+exp(x))
def F1v(x):
    return ( -x/(1+exp(x)) - log(1+exp(-x)) )
def F2v(x):
    return ( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )
# ZT function:
def ZT(Seeb,Sig,KappaE,KappaP):
    return (Seeb**2 * Sig)/(KappaE + KappaP)

# Create sample points within [-10,10]:
npoint  = 101
eta     = linspace(-10, 10, npoint) 

# TE quantities:
S_c     = -F1c(eta) / F0c(eta)
S_v     = -F1v(eta) / F0v(eta)
Sigma_c =  F0c(eta)
Sigma_v =  F0v(eta)
kappa_c =  F2c(eta) - ((F1c(eta))**2 / F0c(eta))
kappa_v =  F2v(eta) - ((F1v(eta))**2 / F0v(eta))

Sigma   =  Sigma_c + Sigma_v
S       = (Sigma_c * S_c + Sigma_v * S_v) / Sigma
kappae  = ( (kappa_c + kappa_v)
          + (Sigma_c * Sigma_v)*(S_c - S_v)**2 / (Sigma) )

#----------#
# PLOTTING #
#----------#

# Example line formats are:
# 'b-'  = blue solid line
# 'r--' = red dashed line
# 'g:   = green dotted line
# if black is needed, use 'k' option (as in cmyk color)

#-----------------------#
# SEEBECK'S COEFFICIENT #
#-----------------------#

plt.figure(figsize=(4,3))
plt.plot(eta, S_c, 'r--',
         eta, S_v, 'g:',
         eta, S, 'b-')
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$S/S_0$')
plt.tight_layout()

# Save figure in PDF file
plt.savefig('gaplessSeebeck.pdf')

#-------------------------#
# ELECTRICAL CONDUCTIVITY #
#-------------------------#

plt.figure(figsize=(4,3))
plt.plot(eta, Sigma_c, 'r--',
         eta, Sigma_v, 'g:',
         eta, Sigma, 'b-')
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$\sigma/\sigma_0$')
plt.tight_layout()

# Save figure in PDF file
plt.savefig('gaplessSigma.pdf')

#-----------------#
# FIGURE OF MERIT #
#-----------------#

plt.figure(figsize=(4,2))
plt.plot(eta, ZT(S, Sigma, kappae, 0), 'b-')
plt.axis([-4,4,0,0.01]) # the format is [xmin,xmax,ymin,ymax]
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$ZT$')
plt.tight_layout()

# Save figure in PDF file
plt.savefig('ZTgapless2band.pdf')
```

## Gapped 2D Dirac material

A gapped 2D Dirac material with the band gap of $2\Delta$ can be described by the Hamiltonian $\mathcal{H} = \mathbf{E} \cdot \mathbf{\sigma}$, where $\mathbf{E} =  (\hbar v_F k_x, \hbar v_F k_y, \Delta)$ and $\sigma$ is the Pauli matrix.  The energy dispersion obtained from this Hamiltonian is $E = \pm |\mathbf{E}| = \pm \sqrt{\hbar v_F |\mathbf{k}|^2 + \Delta^2}$.  The longitudinal velocity is given by $v_x^2 = \frac{v_F^2}{2}\left(\frac{E^2 - \Delta^2}{E^2}\right)$.  Therefore, unlike for the gapless 2D Dirac material, the TDF for the gapped 2D Dirac material is now energy dependent:
\begin{align}
\mathcal{T}(E) &= \frac{C v_F^2}{2} \left(\frac{E^2 - \Delta^2}{E^2} \right),
\end{align}
with the energy range within $[-\infty, -\Delta]$ and $[\Delta, \infty]$ to be considered in the calculation of TE kernel.

### Gapped one band model

In the one band model, we consider a conduction band with energy range $[\Delta, \infty]$ for the gapped 2D Dirac material. Here we will use the dimensionless gap $\tilde{\Delta} = \Delta/(k_B T)$ for simplifying the formulation.  The TE kernel for the conduction band is: 
\begin{align}
\mathcal{L}_{i,c} &= \int_\Delta^\infty \frac{C v_F^2(E^2-\Delta^2)}{2E^2}
(E-\mu)^i\frac{1}{k_BT} \frac{e^{(E-\mu)/(k_B T)}}{(e^{(E-\mu)/(k_B T)} +1 )^2} dE, \\
&= \frac{Cv_F^2(k_BT)^i}{2} \int_{\tilde{\Delta}}^\infty
\frac{(\epsilon^2-\tilde{\Delta}^2)}{\varepsilon^2} (\varepsilon-\eta)^i
\frac{e^{(\epsilon-\eta)}}{(e^{(\epsilon-\eta)}+1)^2} d\varepsilon \\
&= \frac{Cv_F^2(k_BT)^i}{2}\int_{\tilde{\Delta}-\eta}^\infty
\left(1-\frac{\tilde{\Delta}^2}{(x+\eta)^2}\right) \frac{x^i e^x}{(e^x+1)^2} dx,\\
&= \frac{Cv_F^2(k_BT)^i}{2} \left(
\mathcal{F}_{i,c}(\eta-\tilde{\Delta})-\mathcal{G}_{i,c}(\eta, \tilde{\Delta})\right),
\end{align}
with
\begin{align}
\mathcal{G}_{i,c}(\eta, \tilde{\Delta})
= \int_{\tilde\Delta-\eta}^\infty \frac{\tilde\Delta^2}{(x+\eta)^2}
\frac{x^i e^x}{(e^x+1)^2} dx.
\end{align}
This $\mathcal{G}_i$ integral is calculated numerically because it is difficult to solve the integral analytically.

### Code: gapped one band

```python
##################################################
# ZT of GAPPED 2D Dirac material: ONE BAND model #
##################################################

#-----------#
# MAIN PART #
#-----------#

# Define kernel:

delta    = 1    # band gap
infinity = 100  # this value is sufficient for our function

def F0(x):
    return exp(x)/(1+exp(x))
def F1(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )

def G0(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I
def G1(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I
def G2(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I

# ZT function:
def ZT(Seeb,Sig,KappaE,KappaP):
    return (Seeb**2 * Sig)/(KappaE + KappaP)

# Create sample points within [-10,10]:
npoint  = 101
eta     = linspace(-10, 10, npoint) 

L0      = zeros(npoint,float)
L1      = zeros(npoint,float)
L2      = zeros(npoint,float)

S_c     = zeros(npoint,float)
Sigma_c = zeros(npoint,float)
kappa_c = zeros(npoint,float)

for i in range(eta.size):
    L0[i] = F0(eta[i] - delta) - delta**2 * G0(eta[i],delta-eta[i])
    L1[i] = F1(eta[i] - delta) - delta**2 * G1(eta[i],delta-eta[i])
    L2[i] = F2(eta[i] - delta) - delta**2 * G2(eta[i],delta-eta[i])

# TE quantities:
S_c     = -L1 / L0
Sigma_c =  L0
kappa_c =  L2 - (L1**2 / L0)
    
#----------#
# PLOTTING #
#----------#

plt.figure(figsize=(4,3))
plt.plot(eta, ZT(S_c, Sigma_c, kappa_c, 0), 'b-',
         eta, ZT(S_c, Sigma_c, kappa_c, 1), 'r--',
         eta, ZT(S_c, Sigma_c, kappa_c, 2), 'g:')
plt.axis([-4,4,0,2]) # the format is [xmin,xmax,ymin,ymax]
# in plt.text, the option r before the LaTeX is necessary 
plt.text(-0.7,1.8, r'$r_\kappa = 0$', color='b')
plt.text(-0.7,1.0, r'$r_\kappa = 1$', color='r')
plt.text(-0.7,0.2, r'$r_\kappa = 2$', color='g')
# in plt.xlabel or ylabel, the option r is not necessary
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$ZT$')
# Hide the right and top spines (axis)
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Remove whitespaces around the figure
plt.tight_layout()

# Save figure in PDF file
plt.savefig('ZTgapped1band.pdf')
```

### Gapped two band model

In the two band model, we include an additional term for the TE kernel related to a valence band:
\begin{align}
\mathcal{L}_{i,v} &= \int_{-\infty}^\Delta \frac{C v_F^2(E^2-\Delta^2)}{2E^2}
(E-\mu)^i\frac{1}{k_BT} \frac{e^{(E-\mu)/(k_B T)}}{(e^{(E-\mu)/(k_B T)} +1 )^2} dE, \\
&= \frac{Cv_F^2(k_BT)^i}{2} \left(
\mathcal{F}_{i,v}(\eta+\tilde{\Delta})-\mathcal{G}_{i,v}(\eta,\tilde{\Delta})\right),
\end{align}
with
\begin{align}
\mathcal{G}_{i,v}(\eta,\tilde{\Delta})
= \int_{-\infty}^{-\tilde{\Delta}-\eta} \frac{\tilde\Delta^2}{(x+\eta)^2}
\frac{x^i e^x}{(e^x+1)^2} dx.
\end{align}

### Code: gapped two band

```python
##################################################
# ZT of GAPPED 2D Dirac material: TWO BAND model #
##################################################

#-----------#
# MAIN PART #
#-----------#

# Define kernel:

delta    = 1    # band gap
infinity = 100  # this value is sufficient for our function

## conduction band
def F0c(x):
    return exp(x)/(1+exp(x))
def F1c(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2c(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )
def G0c(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I
def G1c(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I
def G2c(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D, infinity, args=(xx))
    return I

## valence band
def F0v(x):
    return 1/(1+exp(x))
def F1v(x):
    return ( -x/(1+exp(x)) - log(1+exp(-x)) )
def F2v(x):
    return ( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )
def G0v(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D, args=(xx))
    return I
def G1v(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D, args=(xx))
    return I
def G2v(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D, args=(xx))
    return I

# ZT function:
def ZT(Seeb,Sig,KappaE,KappaP):
    return (Seeb**2 * Sigma)/(KappaE + KappaP)

# Create sample points within [-10,10]:
npoint  = 101
eta     = linspace(-10, 10, npoint) 

L0c     = zeros(npoint,float)
L1c     = zeros(npoint,float)
L2c     = zeros(npoint,float)

L0v     = zeros(npoint,float)
L1v     = zeros(npoint,float)
L2v     = zeros(npoint,float)

S_c     = zeros(npoint,float)
Sigma_c = zeros(npoint,float)
kappa_c = zeros(npoint,float)

S_v     = zeros(npoint,float)
Sigma_v = zeros(npoint,float)
kappa_v = zeros(npoint,float)

S       = zeros(npoint,float)
Sigma   = zeros(npoint,float)
kappae  = zeros(npoint,float)

for i in range(eta.size):
    L0c[i] = F0c(eta[i] - delta) - delta**2 * G0c(eta[i],delta-eta[i])
    L1c[i] = F1c(eta[i] - delta) - delta**2 * G1c(eta[i],delta-eta[i])
    L2c[i] = F2c(eta[i] - delta) - delta**2 * G2c(eta[i],delta-eta[i])
    L0v[i] = F0v(eta[i] + delta) - delta**2 * G0v(eta[i],delta+eta[i])
    L1v[i] = F1v(eta[i] + delta) - delta**2 * G1v(eta[i],delta+eta[i])
    L2v[i] = F2v(eta[i] + delta) - delta**2 * G2v(eta[i],delta+eta[i])   
    
# TE quantities:
S_c     = -L1c / L0c
S_v     = -L1v / L0v
Sigma_c =  L0c
Sigma_v =  L0v
kappa_c =  L2c - (L1c**2 / L0c)
kappa_v =  L2v - (L1v**2 / L0v)

Sigma   =  Sigma_c + Sigma_v
S       = (Sigma_c * S_c + Sigma_v * S_v) / Sigma
kappae  = ( (kappa_c + kappa_v)
          + (Sigma_c * Sigma_v)*(S_c - S_v)**2 / (Sigma) )

#----------#
# PLOTTING #
#----------#

# Example line formats are:
# 'b-'  = blue solid line
# 'r--' = red dashed line
# 'g:   = green dotted line
# if black is needed, use 'k' option (as in cmyk color)

#-----------------------#
# SEEBECK'S COEFFICIENT #
#-----------------------#

plt.figure(figsize=(4,3))
plt.plot(eta, S_c, 'r--',
         eta, S_v, 'g:',
         eta, S, 'b-')
plt.axis([-8,8,-4,4]) # the format is [xmin,xmax,ymin,ymax]
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$S/S_0$')
plt.tight_layout()

# Save figure in PDF file
plt.savefig('gappedSeebeck.pdf')

#-------------------------#
# ELECTRICAL CONDUCTIVITY #
#-------------------------#

plt.figure(figsize=(4,3))
plt.plot(eta, Sigma_c, 'r--',
         eta, Sigma_v, 'g:',
         eta, Sigma, 'b-')
plt.axis([-8,8,0,1]) # the format is [xmin,xmax,ymin,ymax]
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$\sigma/\sigma_0$')
plt.tight_layout()

# Save figure in PDF file
plt.savefig('gappedSigma.pdf')

#-----------------#
# FIGURE OF MERIT #
#-----------------#

plt.figure(figsize=(4,3))
plt.plot(eta, ZT(S, Sigma, kappae, 0), 'b-',
         eta, ZT(S, Sigma, kappae, 1), 'r--',
         eta, ZT(S, Sigma, kappae, 2), 'g:')
plt.axis([-8,8,0,0.3]) # the format is [xmin,xmax,ymin,ymax]
plt.text(4,0.20, r'$r_\kappa = 0$', color='b')
plt.text(4,0.12, r'$r_\kappa = 1$', color='r')
plt.text(4,0.05, r'$r_\kappa = 2$', color='g')
plt.xlabel('$\mu/k_B T$')
plt.ylabel('$ZT$')
plt.tight_layout()

# Save figure in PDF file
plt.savefig('ZTgapped2band.pdf')
```

## Fig. 3: TE properties of gapped Dirac

```python
##################################
# TE of GAPPED 2D Dirac material #
##################################

#----------#
# PREAMBLE #
#----------#

# The preamble here is the same as the previous figure
# Make sure that we plot the previous figure before

# However, the following size control needs to be repeated
# It's not clear why, but anyway just copy it

SMALL_SIZE  = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)        # controls default text sizes
plt.rc('legend', fontsize=SMALL_SIZE)  # legend fontsize
plt.rc('xtick', labelsize=MEDIUM_SIZE) # fontsize of the x tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE) # fontsize of the y tick labels
plt.rc('axes', labelsize=BIGGER_SIZE)  # fontsize of the x and y labels

#-----------#
# MAIN PART #
#-----------#

# Define kernel:

infinity = 200  # this value is sufficient for our function

## conduction band

def F0c(x):
    return exp(x)/(1+exp(x))
def F1c(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2c(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )

def G0c(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D-xx, infinity, args=(xx))
    return I
def G1c(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D-xx, infinity, args=(xx))
    return I
def G2c(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D-xx, infinity, args=(xx))
    return I

def L0c(Eta,DD):
    return F0c(Eta - DD) - DD**2 * G0c(Eta,DD)
def L1c(Eta,DD):
    return F1c(Eta - DD) - DD**2 * G1c(Eta,DD)
def L2c(Eta,DD):
    return F2c(Eta - DD) - DD**2 * G2c(Eta,DD)

## valence band

def F0v(x):
    return 1/(1+exp(x))
def F1v(x):
    return ( -x/(1+exp(x)) - log(1+exp(-x)) )
def F2v(x):
    return ( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )

def G0v(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D-xx, args=(xx))
    return I
def G1v(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D-xx, args=(xx))
    return I
def G2v(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D-xx, args=(xx))
    return I

def L0v(Eta,DD):
    return F0v(Eta + DD) - DD**2 * G0v(Eta,DD)
def L1v(Eta,DD):
    return F1v(Eta + DD) - DD**2 * G1v(Eta,DD)
def L2v(Eta,DD):
    return F2v(Eta + DD) - DD**2 * G2v(Eta,DD)   

# TE quantities:

def Sc(Eta,DD):
    return -L1c(Eta,DD) / L0c(Eta,DD)
def Sv(Eta,DD):
    return -L1v(Eta,DD) / L0v(Eta,DD)
def sigmac(Eta,DD):
    return L0c(Eta,DD)
def sigmav(Eta,DD):
    return L0v(Eta,DD)
def kappac(Eta,DD):
    return L2c(Eta,DD) - ((L1c(Eta,DD))**2 / L0c(Eta,DD))
def kappav(Eta,DD):
    return L2v(Eta,DD) - ((L1v(Eta,DD))**2 / L0v(Eta,DD))

def sigma(Eta,DD):
    return sigmac(Eta,DD) + sigmav(Eta,DD)
def Seebeck(Eta,DD):
    return ((sigmac(Eta,DD) * Sc(Eta,DD) +
            sigmav(Eta,DD) * Sv(Eta,DD)) / sigma(Eta,DD))
def kappae(Eta,DD):
    return (kappac(Eta,DD) + kappav(Eta,DD) + 
            (sigmac(Eta,DD) * sigmav(Eta,DD))*
            (Sc(Eta,DD) - Sv(Eta,DD))**2 / sigma(Eta,DD))
def ZT(Seeb,Sig,KappaEl,KappaPh):
    return (Seeb**2 * Sig)/(KappaEl + KappaPh)

# Create sample points within [-10,10]:
npoint  = 101
eta     = linspace(-10, 10, npoint) 
delta   = 1

S       = zeros(npoint,float)
S_c     = zeros(npoint,float)
S_v     = zeros(npoint,float)
Sigmae  = zeros(npoint,float)
Sigma_c = zeros(npoint,float)
Sigma_v = zeros(npoint,float)
Kappae  = zeros(npoint,float)
Kappa_c = zeros(npoint,float)
Kappa_v = zeros(npoint,float)

for i in range(eta.size):
    S_c[i]     = Sc(eta[i],delta)
    S_v[i]     = Sv(eta[i],delta)
    S[i]       = Seebeck(eta[i],delta)
    Sigma_c[i] = sigmac(eta[i],delta)
    Sigma_v[i] = sigmav(eta[i],delta)
    Sigmae[i]  = sigma(eta[i],delta)
    Kappae[i]  = kappae(eta[i],delta)
    Kappa_c[i] = kappac(eta[i],delta)
    Kappa_v[i] = kappav(eta[i],delta)

# Define the size of the entire plot
plt.figure(figsize=(6,4))

# Subplot (a): Seebeck's coefficient
ax1 = plt.subplot(2,2,1)
plt.plot(eta, S_c, 'r--')
plt.plot(eta, S_v, 'g:')
plt.plot(eta, S, 'b-',linewidth=2)
plt.axis([-8,8,-4,4]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-8,9,4))
ax1.xaxis.set_minor_locator(AutoMinorLocator(4))
plt.yticks(range(-4,5,2))
ax1.yaxis.set_minor_locator(AutoMinorLocator(4))
ax1.yaxis.labelpad = -4
plt.ylabel('$S$ $(S_0)$')
plt.text(1.8,-2.6, r'$S_c$', color='red')
plt.text(-2.8,2, r'$S_v$', color='green')
plt.text(0,0.4, r'$S$', color='blue')
plt.text(3.7,2.7, r'$\tilde{\Delta} = 1$')
plt.text(-13,3.5,'(a)',fontsize='16')

# Subplot (b): Electrical conductivity
ax2 = plt.subplot(2,2,2)
plt.plot(eta, Sigma_c, 'r--')
plt.plot(eta, Sigma_v, 'g:')
plt.plot(eta, Sigmae, 'b-', linewidth=2)
plt.axis([-8,8,0,1.2]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-8,9,4))
ax2.xaxis.set_minor_locator(AutoMinorLocator(4))
plt.yticks(arange(0,1.2,step=0.5))
ax2.yaxis.set_minor_locator(AutoMinorLocator(5))
plt.ylabel('$\sigma$ $(\sigma_0)$')
plt.text(1.5,0.3, r'$\sigma_c$', color='red')
plt.text(-3,0.3, r'$\sigma_v$', color='green')
plt.text(-0.5,0.5, r'$\sigma$', color='blue')
#plt.text(3.7,1, r'$\tilde{\Delta} = 1$')
plt.text(-14,1.1,'(b)',fontsize='16')

# Subplot (c): Thermal conductivity (electron part)
ax3 = plt.subplot(2,2,3)
plt.plot(eta, Kappa_c, 'r--')
plt.plot(eta, Kappa_v, 'g:')
plt.plot(eta, Kappae, 'b-', linewidth=2)
plt.axis([-8,8,0,3]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-8,9,4))
ax3.xaxis.set_minor_locator(AutoMinorLocator(4))
plt.yticks(range(0,5,1))
ax3.yaxis.set_minor_locator(AutoMinorLocator(4))
ax3.yaxis.labelpad = 10
plt.xlabel('$\mu$ $(k_B T)$')
plt.ylabel('$\kappa_e$ $(\kappa_0)$')
plt.text(3,1, r'$\kappa_{e,c}$', color='red')
plt.text(-5.5,1, r'$\kappa_{e,v}$', color='green')
plt.text(-0.8,2.9, r'$\kappa_e$', color='blue')
#plt.text(3.7,3.3, r'$\tilde{\Delta} = 1$')
plt.text(-13,3.8,'(c)',fontsize='16')

# Subplot (d): Figure of merit
ax4 = plt.subplot(2,2,4)
plt.plot(eta, ZT(S, Sigmae, Kappae, 0), 'C3-', linewidth=2)
plt.plot(eta, ZT(S, Sigmae, Kappae, 1), 'C4-', linewidth=2)
plt.plot(eta, ZT(S, Sigmae, Kappae, 2), 'C5-', linewidth=2)
plt.axis([-8,8,0,0.3]) # the format is [xmin,xmax,ymin,ymax]
plt.xticks(range(-8,9,4))
ax4.xaxis.set_minor_locator(AutoMinorLocator(4))
plt.yticks(arange(0,0.31,step=0.1))
ax4.yaxis.set_minor_locator(AutoMinorLocator(4))
plt.text(-7.6,0.16, r'$r_\kappa = 0$', color='C3')
plt.text(-7.6,0.12, r'$r_\kappa = 1$', color='C4')
plt.text(-7.6,0.08, r'$r_\kappa = 2$', color='C5')
#plt.text(3.7,0.25, r'$\tilde{\Delta} = 1$')
plt.text(-14,0.28,'(d)',fontsize='16')
plt.xlabel('$\mu$ $(k_B T)$')
plt.ylabel('$ZT$')

# Save figure
plt.tight_layout(pad=0.1, w_pad=0.5, h_pad=0.1)
plt.savefig('figTEgapped.pdf')
```

```python
##################################
# TE of GAPPED 2D Dirac material #
##################################

#----------#
# PREAMBLE #
#----------#

# Call necessary mathematical and numerical libraries
from numpy import frompyfunc, linspace, exp, log, pi, zeros, arange, argmax
from mpmath import polylog
from scipy.integrate import quad
from matplotlib.ticker import AutoMinorLocator

# Call plotting library
import matplotlib.pyplot as plt
%matplotlib inline



SMALL_SIZE  = 14
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)        # controls default text sizes
plt.rc('legend', fontsize=SMALL_SIZE)  # legend fontsize
plt.rc('xtick', labelsize=MEDIUM_SIZE) # fontsize of the x tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE) # fontsize of the y tick labels
plt.rc('axes', labelsize=BIGGER_SIZE)  # fontsize of the x and y labels

# Specific treatment for polylogarithm function
plog = frompyfunc(polylog, 2, 1)
# numpy.frompyfunc(func, nin, nout)
# Takes an arbitrary Python function and returns a NumPy ufunc.
# Parameters:
#            func : An arbitrary Python function.
#            nin  : (int) The number of input arguments.
#            nout : (int) The number of objects returned by func.
# Returns a NumPy universal function (ufunc) object.

#-----------#
# MAIN PART #
#-----------#

# Define kernel:

infinity = 100  # this value is sufficient for our function

## conduction band

def F0c(x):
    return exp(x)/(1+exp(x))
def F1c(x):
    return ( x/(1+exp(x)) + log(1+exp(-x)) )
def F2c(x):
    return ( (pi**2)/3 - (x**2 / (1+exp(x)) )
            - 2*x*log(1+exp(-x)) + 2*plog(2,-exp(-x)) )

def G0c(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D-xx, infinity, args=(xx))
    return I
def G1c(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D-xx, infinity, args=(xx))
    return I
def G2c(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, D-xx, infinity, args=(xx))
    return I

def L0c(Eta,DD):
    return F0c(Eta - DD) - DD**2 * G0c(Eta,DD)
def L1c(Eta,DD):
    return F1c(Eta - DD) - DD**2 * G1c(Eta,DD)
def L2c(Eta,DD):
    return F2c(Eta - DD) - DD**2 * G2c(Eta,DD)

## valence band

def F0v(x):
    return 1/(1+exp(x))
def F1v(x):
    return ( -x/(1+exp(x)) - log(1+exp(-x)) )
def F2v(x):
    return ( (x**2 / (1+exp(x)) ) + 2*x*log(1+exp(-x))
            - 2*plog(2,-exp(-x)) )

def G0v(xx,D):
    def integrand(x, xx):
        return (1 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D-xx, args=(xx))
    return I
def G1v(xx,D):
    def integrand(x, xx):
        return (x / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D-xx, args=(xx))
    return I
def G2v(xx,D):
    def integrand(x, xx):
        return (x**2 / (x + xx)**2) * (exp(x) / (exp(x)+1)**2)
    I, error = quad(integrand, -infinity, -D-xx, args=(xx))
    return I

def L0v(Eta,DD):
    return F0v(Eta + DD) - DD**2 * G0v(Eta,DD)
def L1v(Eta,DD):
    return F1v(Eta + DD) - DD**2 * G1v(Eta,DD)
def L2v(Eta,DD):
    return F2v(Eta + DD) - DD**2 * G2v(Eta,DD)   

# TE quantities:

def Sc(Eta,DD):
    return -L1c(Eta,DD) / L0c(Eta,DD)
def Sv(Eta,DD):
    return -L1v(Eta,DD) / L0v(Eta,DD)
def sigmac(Eta,DD):
    return L0c(Eta,DD)
def sigmav(Eta,DD):
    return L0v(Eta,DD)
def kappac(Eta,DD):
    return L2c(Eta,DD) - ((L1c(Eta,DD))**2 / L0c(Eta,DD))
def kappav(Eta,DD):
    return L2v(Eta,DD) - ((L1v(Eta,DD))**2 / L0v(Eta,DD))

def sigma(Eta,DD):
    return sigmac(Eta,DD) + sigmav(Eta,DD)
def Seebeck(Eta,DD):
    return ((sigmac(Eta,DD) * Sc(Eta,DD) +
            sigmav(Eta,DD) * Sv(Eta,DD)) / sigma(Eta,DD))
def kappae(Eta,DD):
    return (kappac(Eta,DD) + kappav(Eta,DD) + 
            (sigmac(Eta,DD) * sigmav(Eta,DD))*
            (Sc(Eta,DD) - Sv(Eta,DD))**2 / sigma(Eta,DD))
def ZT(Seeb,Sig,KappaEl,KappaPh):
    return (Seeb**2 * Sig)/(KappaEl + KappaPh)

# Create sample points:
neta   = 401
ndelta = 41 #81
nkapp  = 6 #11
eta    = linspace(0,20,neta) 
delta  = linspace(0,20,ndelta)
rkapp  = linspace(0,5,nkapp)

# exception for delta = 0 and rkapp = 0
# due to numerical issue
delta[0] = 0.002
rkapp[0] = 0.002

# Initialize TE properties
S      = zeros((neta,ndelta))
Sigmae = zeros((neta,ndelta))
Kappae = zeros((neta,ndelta))
ZTcal  = zeros((neta,ndelta,nkapp))

# Calculate TE properties within the sample points
for i in range(eta.size):
    for j in range(delta.size):
        S[i,j]      = Seebeck(eta[i],delta[j])
        Sigmae[i,j] = sigma(eta[i],delta[j])
        Kappae[i,j] = kappae(eta[i],delta[j])
        for k in range(rkapp.size):
            ZTcal[i,j,k] = ZT(S[i,j],Sigmae[i,j],Kappae[i,j],rkapp[k])

# Finding maximum
ZTopt = zeros((ndelta,nkapp))
etaopt = zeros((ndelta,nkapp))
ZTeta = zeros(neta,float)
for j in range(delta.size):
    for k in range(rkapp.size):
        for i in range(eta.size):
            ZTeta[i] = ZTcal[i,j,k]
        ZTopt[j,k] = max(ZTeta)
        etaopt[j,k]=abs(eta[argmax(ZTeta)])
```
